```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parámetros
alpha = 0.10
alpha_2 = alpha / 2
z_crit = 1.645  # valor z tal que P(Z > 1.645) = 0.05 en la Normal estándar

# Generamos un rango de valores para la variable 'z'
x = np.linspace(-3, 3, 500)
y = norm.pdf(x)  # f(z) = pdf de la Normal(0,1)

# Creamos la figura
plt.figure(figsize=(8, 5))

# Dibujamos la curva de densidad
plt.plot(x, y, color='blue', label='Densidad $N(0,1)$')

# Sombreamos las colas:
#   - Cola izquierda: z < -z_crit
#   - Cola derecha: z > +z_crit
x_left = np.linspace(-3, -z_crit, 200)
y_left = norm.pdf(x_left)
plt.fill_between(x_left, y_left, color='cyan', alpha=0.5)

x_right = np.linspace(z_crit, 3, 200)
y_right = norm.pdf(x_right)
plt.fill_between(x_right, y_right, color='cyan', alpha=0.5)

# Trazamos líneas verticales punteadas en z = ±z_crit
plt.vlines([-z_crit, z_crit], 0, 0.1, color='blue')
plt.axhline(y=0, color="black", linewidth=1.5)

# Etiquetas para z_crit
plt.text(-z_crit, -0.01, r"$-z_{0.95}=-1.645$", 
         ha='center', va='top', fontsize=10, color='blue')
plt.text(z_crit, -0.01, r"$z_{0.95}=1.645$", 
         ha='center', va='top', fontsize=10, color='blue')

# Agregamos un texto indicando alpha/2 en cada cola
# (opcionalmente, con flechas o simplemente un texto cercano)
plt.text(-2.7, 0.06, r"$\frac{\alpha}{2}=0.05$", 
         ha='center', va='bottom', fontsize=9, color='blue')
plt.text(2.7, 0.06, r"$\frac{\alpha}{2}=0.05$", 
         ha='center', va='bottom', fontsize=9, color='blue')

# Etiqueta central que muestra 1 - alpha
plt.text(0, 0.25, r"$1 - \alpha = 0.90$", color="blue", 
         ha='center', va='bottom', fontsize=11)

# Etiquetamos la función
plt.text(0, 0.42, r"$\mathcal{N}(0,1)$",
         ha='center', va='bottom', fontsize=11, color='blue')

# Ajustes finales del gráfico
plt.ylim(0, 0.45)
plt.xlim(-3, 3)
plt.grid(False)
plt.xticks([])  # Quitar valores del eje x
plt.yticks([])  # Quitar valores del eje y
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['bottom'].set_visible(False)
plt.gca().spines['left'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().set_facecolor("none")
plt.gca().patch.set_facecolor("none")
plt.gcf().patch.set_facecolor("none")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t

# Parámetros
alpha = 0.10
alpha_2 = alpha / 2
df = 10  # Grados de libertad para t-Student
t_crit = t.ppf(1 - alpha_2, df)  # Valor crítico t para alpha/2

# Generamos un rango de valores para la distribución t
x = np.linspace(-4, 4, 500)
y = t.pdf(x, df)  # Función de densidad de la distribución t

# Creamos la figura
plt.figure(figsize=(8, 5))

# Dibujamos la curva de densidad
plt.plot(x, y, color='blue', label='Distribución $t$ de Student')

# Sombreamos el área central entre -t_crit y t_crit
x_fill_1 = np.linspace(-4, -t_crit, 200)
y_fill_1 = t.pdf(x_fill_1, df)
plt.fill_between(x_fill_1, y_fill_1, color='blue', alpha=0.2)

x_fill_2 = np.linspace(t_crit, 4, 200)
y_fill_2 = t.pdf(x_fill_2, df)
plt.fill_between(x_fill_2, y_fill_2, color='blue', alpha=0.2)

# Trazamos líneas verticales punteadas en ±t_crit
plt.vlines([-t_crit, t_crit], 0, 0.08, color='blue')
plt.vlines(0, 0, 0.39, color="blue")
# plt.hlines(y=0.08, xmin=-t_crit, xmax=t_crit, color="blue", linestyle="--")

# Etiquetas en -t_crit y t_crit
plt.text(-t_crit, -0.02, r"$-t_{n_A+n_B-2}, 1-\frac{\alpha}{2}}$", 
         ha='center', fontsize=10, color='blue')
plt.text(t_crit, -0.02, r"$t_{n_A+n_B-2}, 1-\frac{\alpha}{2}}$", 
         ha='center', fontsize=10, color='blue')

# Etiqueta central para 1-alpha
plt.text(0, 0.2, r"$1-\alpha$", ha='center', fontsize=12, color="blue")

# Etiquetas en las colas para alpha/2
plt.text(-3.5, 0.05, r"$-\frac{\alpha}{2}$", ha='center', fontsize=10, color='blue')
plt.text(3.5, 0.05, r"$\frac{\alpha}{2}$", ha='center', fontsize=10, color='blue')

# Etiqueta del ancho del intervalo
# plt.annotate("", xy=(-4, -0.05), xytext=(4, -0.05),
#              arrowprops=dict(arrowstyle='<->', color='black'))
# plt.text(0, -0.1, r"$1-\frac{\alpha}{2}$", ha='center', fontsize=10)

# Ajustes finales del gráfico
plt.ylim(-0.15, 0.4)
plt.xlim(-4, 4)
plt.xticks([])  # Sin valores en el eje x
plt.yticks([])  # Sin valores en el eje y
plt.axhline(0, color='black', linewidth=0.8)  # Eje horizontal
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['bottom'].set_visible(False)
plt.gca().spines['left'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().set_facecolor("none")
plt.gca().patch.set_facecolor("none")
plt.gcf().patch.set_facecolor("none")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

# Parámetros
alpha = 0.10
alpha_2 = alpha / 2
df = 10  # Grados de libertad para chi-cuadrado

# Valores críticos para la distribución chi-cuadrado
chi2_lower = chi2.ppf(alpha_2, df)  # Valor crítico inferior
chi2_upper = chi2.ppf(1 - alpha_2, df)  # Valor crítico superior

# Generamos un rango de valores para la distribución chi-cuadrado
x = np.linspace(0, chi2.ppf(0.995, df), 500)
y = chi2.pdf(x, df)  # Función de densidad de la distribución chi-cuadrado

# Creamos la figura
plt.figure(figsize=(8, 5))

# Dibujamos la curva de densidad
plt.plot(x, y, color='blue', label=r'Distribución $\chi^2$')

# Sombreamos el área central entre chi2_lower y chi2_upper
x_fill = np.linspace(chi2_upper, 25, 200)
y_fill = chi2.pdf(x_fill, df)
plt.fill_between(x_fill, y_fill, color='cyan', alpha=0.5)
x_fill_2 = np.linspace(0, chi2_lower, 200)
y_fill_2 = chi2.pdf(x_fill_2, df)
plt.fill_between(x_fill_2, y_fill_2, color='cyan', alpha=0.5)
plt.vlines(chi2_upper, 0, 0.015, color='blue', linestyle='--')
plt.vlines(chi2_lower, 0, 0.0425, color='blue', linestyle='--')

# Etiquetas en los valores críticos
plt.text(chi2_lower, -0.01, r"$\chi^2_{n-1, \frac{\alpha}{2}}$", 
         ha='center', fontsize=10, color='blue')
plt.text(chi2_upper, -0.01, r"$\chi^2_{n-1, 1-\frac{\alpha}{2}}$", 
         ha='center', fontsize=10, color='blue')

# Etiqueta central para 1-alpha
plt.text((chi2_lower + chi2_upper) / 2, max(y), r"$1-\alpha$", color="blue",
         ha='center', fontsize=12)

plt.text(2, 0.02, r"$-\frac{\alpha}{2}$", ha='center', fontsize=10, color='blue')
plt.text(22.5, 0.02, r"$\frac{\alpha}{2}$", ha='center', fontsize=10, color='blue')

# Ajustes finales
plt.xticks([])  # Sin valores en el eje x
plt.yticks([]) 
plt.ylim(0, 0.1)
plt.xlim(-1, 25)
plt.axhline(0, color="black", linewidth=1.5)
plt.axvline(0, color="black", linewidth=1.5)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['bottom'].set_visible(False)
plt.gca().spines['left'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().set_facecolor("none")
plt.gca().patch.set_facecolor("none")
plt.gcf().patch.set_facecolor("none")
```
