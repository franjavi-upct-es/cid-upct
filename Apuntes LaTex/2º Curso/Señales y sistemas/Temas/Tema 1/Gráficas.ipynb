```python
import matplotlib.pyplot as plt
import numpy as np

# Crear un array de valores de t desde -10 hasta 10 con un paso de 0.1
t = np.arange(-10, 10, 0.1)

# Crear un array de valores de x
x = np.zeros_like(t)

# Para t >= 3, establecer x(t) a e^{-0.66t}
x[t >= 3] = np.exp(-0.66*t[t >= 3])

# Crear el gráfico
plt.figure(figsize = (10,3))
plt.plot(t, x)
plt.title('Gráfico de x(t)')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.grid(True)

plt.xlim([-10, 10])
plt.ylim([0, 0.15])
plt.yticks([0.05, 0.1, 0.15])

plt.show()
```
