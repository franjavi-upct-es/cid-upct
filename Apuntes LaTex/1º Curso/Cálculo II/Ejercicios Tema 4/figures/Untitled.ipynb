```python
import numpy as np
import matplotlib.pyplot as plt

# Definir el rango de x
x1 = np.linspace(-10, 10, 1000)
x2 = np.linspace(-10, 10, 1000)

# Definir las funciones
y1 = abs(x1)
y2 = -abs(x2)

# Crear la gráfica
fig, ax = plt.subplots(figsize=(8, 6))

# Graficar y1
ax.plot(x1, y1, color='#007AFF', linewidth=1.5, label=r'$y=|x|$')

# Graficar y2
ax.plot(x2, y2, color='#0000FF', linewidth=1.5, label=r'$y=-|x|$')

# Configurar límites de los ejes
ax.set_xlim(-4, 4)
ax.set_ylim(-4, 4)

# Centrar los ejes
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')  # Ocultar el lado derecho
ax.spines['top'].set_color('none')    # Ocultar la parte superior

# Añadir ticks en los ejes centrados
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

# Añadir leyenda
ax.legend(loc='upper right', fontsize=12)



# Mostrar la gráfica
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Crear una malla de puntos en coordenadas cilíndricas
theta = np.linspace(0, 2 * np.pi, 100)
r = np.linspace(0, 1, 50)
theta, r = np.meshgrid(theta, r)
x = r * np.cos(theta)
y = r * np.sin(theta)

# Superficie inferior (cono): z = sqrt(x^2 + y^2) = r
z_inferior = r

# Superficie superior: z = 1
z_superior = np.ones_like(r)

# Crear la figura
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Graficar la superficie inferior (cono)
ax.plot_surface(x, y, z_inferior, cmap='viridis', alpha=0.7, edgecolor='none')

# Graficar la superficie superior (plano z=1)
ax.plot_surface(x, y, z_superior, cmap='plasma', alpha=0.7, edgecolor='none')

# Ajustar la vista y los ejes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.view_init(elev=30, azim=45)

# Display the plot
plt.savefig('Figure 6.png')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros del cilindro
theta = np.linspace(0, 2 * np.pi, 100)
z = np.linspace(-1, 1, 50)
theta, z = np.meshgrid(theta, z)

# Ecuación del cilindro x^2 + y^2 + x = 0
# En coordenadas polares: x = r*cos(theta) - 0.5, y = r*sin(theta)
r = 0.5  # Radio del cilindro
x = r * np.cos(theta) - 0.5  # Ajuste del centro a (-0.5, 0)
y = r * np.sin(theta)

# Graficar el cilindro
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, alpha=0.6, rstride=5, cstride=5, cmap='viridis', edgecolor='none')

# Configuración del gráfico
ax.set_title("Cilindro definido por $x^2 + y^2 + x = 0$", fontsize=14)
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.set_xlim([-1, 1])
ax.set_ylim([-1, 1])
ax.set_zlim([-1, 1])

plt.show()
```

```python
import numpy as np
import plotly.graph_objects as go

# Parámetros del cilindro
theta = np.linspace(0, 2 * np.pi, 100)
z = np.linspace(-1, 1, 50)
theta, z = np.meshgrid(theta, z)

# Ecuación del cilindro x^2 + y^2 + x = 0
r = 0.5  # Radio del cilindro
x = r * np.cos(theta) - 0.5  # Ajuste del centro a (-0.5, 0)
y = r * np.sin(theta)

# Crear la superficie del cilindro
cylinder_surface = go.Surface(x=x, y=y, z=z, colorscale='Viridis', opacity=0.8)

# Configurar el gráfico
fig = go.Figure(data=[cylinder_surface])
fig.update_layout(
    title="Cilindro definido por $x^2 + y^2 + x = 0$",
    scene=dict(
        xaxis_title='X',
        yaxis_title='Y',
        zaxis_title='Z',
        xaxis=dict(range=[-1.5, 1]),
        yaxis=dict(range=[-1, 1]),
        zaxis=dict(range=[-1, 1]),
    ),
)

# Mostrar el gráfico interactivo
fig.show()
```
