```python
import numpy as np
import matplotlib.pyplot as plt

# Definir el rango de x
x1 = np.linspace(0, 8, 1000)  # Intervalo para y = sqrt(2x)
x2 = np.linspace(-4, 4, 1000)  # Intervalo para y = sqrt(8 - 2x)

# Definir las funciones
y1 = np.sqrt(2 * x1)  # y = sqrt(2x)
y2 = np.sqrt(8 - 2 * x2)  # y = sqrt(8 - 2x)

# Crear la gráfica
plt.figure(figsize=(8, 6))

# Graficar y1
plt.plot(x1, y1, color='#007AFF', linewidth=1.5, label=r'$y^2 = 2x$')

# Graficar y2
plt.plot(x2, y2, color='#0000FF', linewidth=1.5, label=r'$y^2 = 8 - 2x$')

# Añadir leyenda
plt.legend(loc='upper right', fontsize=12)

# Configurar límites de los ejes
plt.ylim(-1, 5)
plt.xlim(-4, 8)

# Centrar los ejes
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)

# Añadir etiquetas y título
plt.xlabel('$x$', fontsize=14)
plt.ylabel('$y$', fontsize=14)

# Mostrar la gráfica
plt.grid(False)
plt.show()
```
