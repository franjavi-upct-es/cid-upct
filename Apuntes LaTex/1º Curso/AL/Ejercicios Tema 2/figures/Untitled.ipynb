```python
import matplotlib.pyplot as plt
import numpy as np

norm_1 = lambda x, y: np.abs(x) + np.abs(y)
norm_2 = lambda x, y: np.sqrt(x**2 + y**2)
norm_inf = lambda x, y: np.maximum(np.abs(x), np.abs(y))

x = np.linspace(-1.5, 1.5, 1000)
y = np.linspace(-1.5, 1.5, 1000)
X, Y = np.meshgrid(x, y)

Z1 = norm_1(X, Y)
Z2 = norm_2(X, Y)
Zinf = norm_inf(X, Y)

# Configurar la grfica
plt.figure(figsize=(8, 8))
# Centrar los ejes
plt.gca().spines['left'].set_position('zero')
plt.gca().spines['bottom'].set_position('zero')
plt.gca().spines['right'].set_color('none')  # Ocultar el lado derecho
plt.gca().spines['top'].set_color('none')    # Ocultar la parte superior

# Aadir ticks en los ejes centrados
plt.gca().xaxis.set_ticks_position('bottom')
plt.gca().yaxis.set_ticks_position('left')
# Dibujar norma 1
plt.contour(X, Y, Z1, levels=[1], colors='blue', label=r'$\|(x,y)\|_1=1$')
plt.annotate(r'$\|(x,y)\|_1=1$', xy=(1, 0.05), color='blue', fontsize=12)

# Dibujar norma 2
plt.contour(X, Y, Z2, levels=[1], colors='red', label=r'$\|(x,y)\|_2=1$')
plt.annotate(r'$\|(x,y)\|_2=1$', xy=(0.7, 0.7), color='red', fontsize=12)

# Dibujar norma infinito
plt.contour(X, Y, Zinf, levels=[1], colors='green', label=r'$\|(x,y)\|_\infty=1$')
plt.annotate(r'$\|(x,y)\|_\infty=1$', xy=(1, 1), color='green', fontsize=12)

# Mostrar la grfica
plt.savefig("Figure 1.png")
```
