```python
import numpy as np
import matplotlib.pyplot as plt

# Subset (a): Numbers with modulus 1
theta = np.linspace(0, 2 * np.pi, 500)
x_a = np.cos(theta)
y_a = np.sin(theta)

# Subset (b): Roots of unity (8th roots with positive exponent)
k = np.arange(1, 9)
z_b = np.exp(2j * np.pi * k / 8)
x_b, y_b = z_b.real, z_b.imag

# Subset (c): Roots of unity (8th roots with negative exponent)
z_c = np.exp(-2j * np.pi * k / 8)
x_c, y_c = z_c.real, z_c.imag

# Subset (d): Numbers such that z - conjugate(z) = j
# z = x + yj, z - conjugate(z) = 2yj = j -> y = 1/2
x_d = np.linspace(-2, 2, 500)
y_d = np.full_like(x_d, 0.5)
```

```python
# Plot (a): |z| = 1
plt.figure(figsize=(6, 6))
plt.plot(x_a, y_a, label="|z| = 1", color='blue')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title("Subset (a): |z| = 1")
plt.xlabel("Re(z)")
plt.ylabel("Im(z)")
plt.axis('equal')
plt.legend()
plt.show()

# Plot (b): Positive 8th roots of unity
plt.figure(figsize=(6, 6))
plt.scatter(x_b, y_b, color='red', label=r"$1\le k\le 8$")
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.xlabel("Re(z)")
plt.ylabel("Im(z)")
plt.axis('equal')
plt.legend()
plt.show()

# Plot (c): Negative 8th roots of unity
plt.figure(figsize=(6, 6))
plt.scatter(x_c, y_c, color='green', label=r"$1\le k\le 8$")
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.xlabel("Re(z)")
plt.ylabel("Im(z)")
plt.axis('equal')
plt.legend()
plt.show()

# Plot (d): z - conjugate(z) = j
plt.figure(figsize=(6, 6))
plt.plot(x_d, y_d, color='purple', label="z - conjugate(z) = j")
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.xlabel("Re(z)")
plt.ylabel("Im(z)")
plt.axis('equal')
plt.legend()
plt.show()
```
