Define una función que multiplique y divida $100$ veces a tu número de DNI el 
número $1/10$.
Imprime en pantalla el valor resultante y ejecuta un operador booleano para 
averiguar si el valor obtenido coincide con tu DNI.

¿Obtienes tu DNI? ¿Por qué?

```python
# Completar aquí
def multiplicacion_y_division(iterations):
    result = 26649110
    for i in range(iterations):
        result *= 1 / 10
    for i in range(iterations):
        result /= 1 / 10
    return result
multiplicacion_y_division(100)

#No obtengo mi DNI porque debido a los números decimales y a que el resultado no está redondeado.
```

Consider los números complejos $z_1 = a + bj$ y $z_2 = g + hj$.
Calcula la forma polar de ambos números. Expresa la fase en grados

```python
# Completar aquí
import cmath
a = 2
b = 6
g = 1
h = 3
# los convertimos a forma binómica
z1 = complex(a , b) 
print(type(z1))
z2 = complex(g , h)
print(type(z2))
#  lo convertimos a forma polar (argumentos en radianes)
x = cmath.polar(z1)
print (f"El módulo y argumento de z1 =  {x}")
y = cmath.polar(z2)
print (f"El módulo y argumento de z2 =  {y}")
```

```python
# el ángulo en grados
import numpy as np 
print('Argumento (en grados) de z1 =', np.rad2deg(x[1]))
print('Argumento (en grados) de z2 =', np.rad2deg(y[1]))
```

Introduce el número compelo $z = e^{2\pi j/ 8}$ y calcula
$z^k$ para $1\leq k\leq 9$

¿Es $z = z^9$ ? ¿Por qué?

```python
# Completar aquí
import cmath
z = complex (0 , 2 * cmath.pi /8)
print (cmath.exp(z))
print(f"z**1 = {cmath.exp(z)}")
print(f"z**2 = {cmath.exp(z) ** 2}")
print(f"z**3 = {cmath.exp(z) ** 3}")
print(f"z**4 = {cmath.exp(z) ** 4}")
print(f"z**5 = {cmath.exp(z) ** 5}")
print(f"z**6 = {cmath.exp(z) ** 6}")
print(f"z**7 = {cmath.exp(z) ** 7}")
print(f"z**8 = {cmath.exp(z) ** 8}")
print(f"z**9 = {cmath.exp(z) ** 9}")
print(f"z**11 = {cmath.exp(z) ** 11}")
print(f"z**13 = {cmath.exp(z) ** 13}")

# z = z ^ 9, porque al obtener el resultado de z ^ 3 vemos que es el mismo resultado de z ^ 1 pero con un valor negativo.
# Se puede apreciar que en z ^ 1, z ^ 7 y z ^ 9, obtener el mismo resultado con valor positivo, mientras que en z ^ 3 y z ^ 5 obtenemos los resultados con valor negativo.
# Se pueden comprobar estas sutiaciones mediante la aplicación de z ^ 11 y z ^ 13, donde obtenemos de nuevo valores negativos.
```
