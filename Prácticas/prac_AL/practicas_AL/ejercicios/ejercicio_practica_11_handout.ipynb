## Factorización SVD y aplicación a PCA (Principal Component Analysis)


En este ejercicio practicaremos el uso de la descomposición en valores singulares (SVD) en el llamado Análisis de Componentes Principales (PCA por sus siglas en inglés). De hecho, PCA es tal importancia en **Data Science**  que el módulo de Python, dedicado al **Machine Learning** tiene implementado el método **pca()** para realizar dicho análisis.



Introduce una matriz **A** de tamaño $5\times 7$ con entradas aleatorias



```python
import numpy as np
```

```python
# Completar aquí
A = np.random.randn(5, 7)
print(f"A = \n {A}")
# --------------------
```

Calcula la media de las entradas de cada columna en la matriz $A^T$. Esto se hace con el método **mean()** de **numpy**. Lo puedes ver en [numpy.mean() API](https://docs.scipy.org/doc/numpy-1.13.0/reference/generated/numpy.mean.html)

```python
# Completar aquí
M = np.mean(A, axis=0)
print(M)
# --------------------
```

Mediante una operación tipo **broadcasting** calcula $C = A - M$

```python
# Completar aquí
C = A - M
print(f"C = \n {C}")
# --------------------
```

En Estadística, dada una matriz de datos $C$, se introduce el concepto de **matriz de covarianza** (lo veréis en las asignaturas de Estadística), que trata de medir la correlación de los datos contenidos en columnas distintas. En Python se calcula escribiendo **D = np.cov(C.T)**. Puedes verlo en [numpy.cov() API](https://docs.scipy.org/doc/numpy-1.13.0/reference/generated/numpy.cov.html)

Calcula la matriz de covarianza de la matriz $C$ anterior. Verás que te sale una matriz simétrica.

```python
# Completar aquí
D = np.cov(C.T)
print(f"D = \n {D}")
# --------------------
```

Calcula la factorización **SVD** de la matriz $D$ anterior.



```python
# Completar aquí
from scipy.linalg import svd
U, s, V = svd(D)
print(f"U = \n {U}")
print(f"s = {s}")
print(f"V = \n {V}")
# --------------------
```

A partir de los valores singulares, construye la matriz **Sigma** de la descomposición **SVD**

```python
# Completar aquí
Sigma = np.zeros((D.shape[0], D.shape[1]))
Sigma[:D.shape[0], :D.shape[0]] = np.diag(s)
print(f"Sigma = \n {Sigma}")
# --------------------
```

Extrae la información de los datos originales asociados a los $4$ primeros valores singulares

```python
# Completar aquí
n_componentes = 4
Sigma = Sigma[:, :n_componentes]
print(f"Sigma reducida = \n {Sigma}")

T = U.dot(Sigma)
print(f"T = \n {T}") 
# T reemplaza a la matriz original D pues contiene la escala de los datos de D.
# --------------------
```
