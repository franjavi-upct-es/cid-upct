## Dot Product Similarity Matrix

Una clase de matrices que aparencen de manera recurrente en varias aplicaciones
de Machine Learning son matrices S que toman la forma S = DD^T, donde D es una
matriz de datos de tamaño nxd y D^T es su traspuesta.
En este ejercicio practicaremos el manejo de este tipo de matrices haciendo uso
del módulo Numpy de Python.

Introduce una matriz D de 4 filas y 8 columnas de modo que la primera fila
contenga tu número de DNI y el resto de filas números de DNI de algunos
familiares tuyos.
Imprime también por pantalla la matriz D

```python
import numpy as np
```

```python
# Completar aquí
D = np.array([
    [2,6,6,4,9,1,1,0],
    [3,6,4,7,9,5,1,0],
    [4,5,3,8,8,1,2,6],
    [7,6,4,1,1,9,4,1]
])
print(f"D =\n{D}")
# --------------------
```

Calcula, usando las dos formas estudiadas en clase, e imprime por pantalla
la matriz S = DD^T

```python
# Completar aquí
S = D.dot(D.T)
print(f"S = \n {S}")
# --------------------
```

Introduce un vector b no nulo de dimensión 4, el que quieras, y calcula, de las
dos formas que vimos en clase, el producto de S y b, es decir, S*b.
Calcula también b*S. Compara los resultados de S*b y b*S.
¿Qué explicación encuentras de ambos resultados?

```python
# Completar aquí
b = np.array([1,4,5,3])
print(f"Sb = \n {S * b}")
print(f"bS = \n {b * S}")
# Los resultados son los mismos debido a la propiedad distributiva donde S*b=b*S
# --------------------
```

Calcula las partes triangular inferior y superior de D.

```python
# Completar aquí
d = np.diag(S)
x = np.diag(d)
print(d)
print(x)
# --------------------
```

Calcula la inversa de S, su traza, su determinante, su rango, su norma de Frobenius y su número de condicionamiento

```python
# Completar aquí
print(f"Inversa de S = \n {np.linalg.inv(S)}")
print(f"Traza de S = {np.trace(S)}")
print(f"Determinante de S = {np.linalg.det(S)}")
print(f"Rango de S = {np.linalg.matrix_rank(S)}")
print(f"Norma de Frobenius de S = {np.linalg.norm(S, 'fro')}")
print(f"Número de condicionamiento de S = {np.linalg.cond(S, 'fro') * np.linalg.norm(np.linalg.inv(S), 'fro')}")
# --------------------
```
