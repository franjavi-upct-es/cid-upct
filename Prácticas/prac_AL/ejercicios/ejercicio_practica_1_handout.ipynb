Define una función que multiplique y divida $100$ veces a tu número de DNI el 
número $1/10$.
Imprime en pantalla el valor resultante y ejecuta un operador booleano para 
averiguar si el valor obtenido coincide con tu DNI.

¿Obtienes tu DNI? ¿Por qué?

```python
# Completar aquí
def calc_dni(dni):
    for i in range(1,101):
        prueba = dni / i
        prueba = prueba * i
    return prueba

print(calc_dni(dni))
print(f"¿Coincide el resultado con el DNI?: {dni == calc_dni(dni)}")
# --------------------
# No devuelve el mismo valor ya que no redondea el resultado
print(f"Nuevo intento: {dni == round(calc_dni(dni))}")
# Si usamos la función "round" si son iguales
```

Introduce el número compelo $z = e^{2\pi j/ 8}$ y calcula
$z^k$ para $1\leq k\leq 9$

¿Es $z = z^9$ ? ¿Por qué?

```python
# Completar aquí
e = complex(0, (2*np.pi)/8)
z = np.exp(e)
print(f"z = {z}")
print()
for i in range(1,10):
    print(z ** i)
print()
print(z == z**9)
print()

# No son iguales por pequeñas diferencias decimales.
# Vamos a probar si redondeando funciona

print(np.round(z) == np.round(z**9))

#Redondeando vemos que si son iguales

# --------------------
```
