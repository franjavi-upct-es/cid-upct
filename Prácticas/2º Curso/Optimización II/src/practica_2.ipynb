## Regresión lineal paso a paso (from scrath)

```python
import numpy as np
```

```python
# Generamos los datos
np.random.seed(42)
X = 2 * np.random.rand(100, 1)          # features
y = 4 + 3 * X + np.random.rand(100, 1) # labels
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X, y)
plt.show()
```

```python
X_b = np.c_[np.ones(X.shape[0]), X]
# Resolvemos el sistema de optimalidad

theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)

print(f"theta_best = {theta_best}")
```

```python
# Definimos el modelo de predicción
def model_predict(x):
    y_hat = theta_best[0] + theta_best[1] * x
    return y_hat
```

```python
plt.plot(X, y, 'o', label='datos')
x = np.linspace(0, 2, 100)
y_predict = model_predict(x)
plt.plot(x, y_predict, label='modelo de predicción')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(framealpha=1, shadow=True)
plt.grid(alpha=0.5)
plt.show()
```

### Regresión lineal en Numpy

```python
theta_best_svd, residuals, rank, s = np.linalg.lstsq(X_b.T.dot(X_b), X_b.T.dot(y))
print(f"theta_best_svd = {theta_best_svd}")
```

## Regresión lineal con Scikit-Learn

```python
from sklearn.datasets import make_regression

X_train, y_train = make_regression(n_samples=1000, n_features=10)
```

```python
from sklearn import linear_model

regr = linear_model.LinearRegression()
regr.fit(X_train, y_train)
print(f"Coeficientes = \n{regr.coef_}")
```
