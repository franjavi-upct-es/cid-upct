## Perceptrones monocapa


**Introducci칩n**

Como se ha visto en teor칤a el perceptr칩n monocapa combina de forma lineal los valores del dato de entrada, les suma un "bias" y al resultado le aplica una funci칩n de activaci칩n no lineal. Esta funci칩n toma como valor +1 si el valor de entrada es positivo y -1 si el valor de entrada es negativo. La operaci칩n se muestra en la imagen siguiente:



Los par치metros del perceptr칩pn son los pesos **w** y el *bias (b)*. La operaci칩n matem치tica es:

$$
\begin{equation*}
    z = \textbf{w}^T\textbf{x}  + b
\end{equation*}
$$


**Conjunto de datos**

En el problema de clasificaci칩n vamos a trabajar con el conocido conjunto de datos ("dataset") **Iris**: 칠ste contiene medidas de longitud y ancho de p칠talos y s칠palos de flores de tres especies del g칠nero Iris. El conjunto de datos contiene 150 registros. Vamos a cargar todos los registros, pero para trabajar con el perceptr칩n monocapa vamos a emplear 칰nicamente dos caracter칤sticas de dos clases: la **longitud y anchura del p칠talo** de los registros correspondientes a **versicolor** y **virginica**.


A continuaci칩n, cargamos los datos del conjunto de datos Iris como DataFrame para ver su estructura de forma clara. 

```python
# Vamos a trabajar con la estructura de datos de Iris cuya descripci칩n se muestra en: 
# https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_iris.html

# Importaci칩n de los m칩dulos necesarios de la biblioteca scikit learn
from sklearn import datasets

# Importamos el conjunto de datos como un dataframe para ver la estructura 
caracteristicas,etiquetas = datasets.load_iris(return_X_y = True, as_frame = True)

print('Las caracter칤sticas son:')
print(caracteristicas)
print('')
print('Las etiquetas son:')
print(etiquetas)
```

Seleccionamos el subconjunto de datos deseado.
Versicolor son las etiquetas ("targets") iguales a 1 y virginica las iguales a 2.

```python
# Valores de versicolor y de virginica

clase_A = 1 # VERSICOLOR
clase_B = 2 # VIRGINICA

# Valores de las etiquetas de correspondientes a la longitud y anchura del p칠talo
caracteristicas = caracteristicas[['petal length (cm)', 'petal width (cm)']]

# Valores de las caracter칤sticas de las clases deseadas
X = caracteristicas[ (etiquetas == clase_A) | (etiquetas == clase_B)]
X = X.values

# Valores de las etiquetas de las clases deseadas
y = etiquetas[ (etiquetas == clase_A) | (etiquetas == clase_B)]
y = y.values
```

Visualizamos las dos clases.

```python
# Figura
fig, ax = plt.subplots(figsize = (7, 7))


X_versicolor = X[y == clase_A]
x1 = X_versicolor[:,0]
x2 = X_versicolor[:,1]
colores = '#007AFF'
plt.scatter(x1, x2, color=colores, label='Versicolor (target 1)')


X_virginica = X[y == clase_B]
x1 = X_virginica[:,0]
x2 = X_virginica[:,1]
colores = 'red'
plt.scatter(x1, x2, color=colores, label='Virginica (target 2)')

plt.title("Longitud p칠talo vs. Anchura p칠talo")
plt.xlabel("Longitud p칠talo (cm)")
plt.ylabel("Anchura p칠talo (cm)")

plt.grid(True)
plt.legend()


print(X_versicolor[:,0])
print(X_versicolor[:,1])
```

**Conjuntos de entrenamiento y test**

Ahora vamos a generar los conjuntos de entrenamiento y test: como sabemos el primero servir치 para estimar los par치metros del modelo y el segundo para comprobar la capacidad de generalizaci칩n. Seleccionamos un conjunto de test de tama침o igual al 20% del total.


```python
# Importaci칩n del m칠todo train_test_split
from sklearn.model_selection import train_test_split

# Seleccionamos un 80% de los datos para entrenamiento y un 20% para test
X_ent, X_test, y_ent, y_test  = train_test_split(X, y, test_size=0.2)
```

**Estandarizaci칩n de los datos**

Antes de entrenar el perceptr칩n se aplicar치 una fase de estandarizaci칩n de los datos de entrada. La estandarizaci칩n eliminar치 la media de cada caracter칤stica y escalar치 para que la varianza sea la unidad. De este modo no existir치n diferencias de escala en las caracter칤sticas de entrada, esta operaci칩n mejora el rendimiento de los modelos de machine learning. Para realizar esta tarea emplearemos la clase `StandardScaler()` de Scikit-learn.

En primer lugar importamos la clase y la instanciamos.

```python
# Importamos la clase StandardScaler para realizar el escalado
from sklearn.preprocessing import StandardScaler

# Instanciamos el escalador
scaler = StandardScaler() # NOS PERMITE NORMALIZAR LOS DATOS, QUE TODOS EST칄N EN LA MISMA ESCALA 
```

Hay que calcular la estandarizaci칩n con las caracter칤sticas del conjunto de entrenamiento y aplicar la transformaci칩n generada al conjunto de entrenamiento y de test.

```python
# Estandariza y prepara correctamente los datos para despu칠s entrenar el modelo con el conjunto de entrenamiento
X_ent_escaladas = scaler.fit_transform(X_ent)

# Aplicamos la estandarizaci칩n calculada con los datos de entrenamiento a los datos de test
X_test_escaladas = scaler.transform(X_test)
```

**Entrenamiento**

El entrenamiento del perceptr칩n monocapa admite cambios en los hiperpar치metros. Por ejemplo se puede fijar el n칰mero m치ximo de iteraciones de aplicaci칩n de la regla de entrenamiento del perceptr칩n. Tambi칠n se puede fijar la constante de actualizaci칩n de los pesos de la combinaci칩n lineal. Vamos a utilizar los siguientes valores:

<font color='blue'> N칰mero m치ximo de iteraciones </font>, `max_iter` igual a 500. El valor por defecto es 1000.

<font color='blue'>Constante de actualizaci칩n de los pesos</font>, `eta0` igual a 0.1. El valor por defecto es 1.

En la siguiente celda importamos la clase `Perceptron` y generamos el modelo

```python
# Importaci칩n de la clase Perceptron
from sklearn.linear_model import Perceptron

# Instanciamos la clase Perceptron
perceptron_monocapa = Perceptron(max_iter = 500, eta0 = 0.1)

# Entrenamos el perceptr칩n monocapa: para ello empleamos el m칠todo fit
perceptron_monocapa.fit(X_ent_escaladas, y_ent)

# Obtenemos las predicciones para el conjunto de entrenamiento y de test
y_sal_ent = perceptron_monocapa.predict(X_ent_escaladas)
y_sal_test = perceptron_monocapa.predict(X_test_escaladas)
```

La siguiente funci칩n ser치 empleada para dibujar los datos de entrenamiento y de test y la frontera de decisi칩n.

```python
# Frontera de decisi칩n: definimos una funci칩n para reutilizar el c칩digo
def dibuja_frontera_decision(Entradas, modelo, Salidas, clase_A, clase_B, 
                            colores, etiqueta_leyenda_cero,
                            etiqueta_leyenda_uno, titulo, titulo_eje_x, titulo_eje_y,
                            simbolo_clase_A, simbolo_clase_B, tamano_simbolo, adicion_frontera):
    '''Dibuja las clases y la frontera de decisi칩n generada por la 
    t칠cnica de machine learning'''

    if (adicion_frontera == 1):
        minX1 = min(Entradas[:, 0])       # Valor m칤nimo de la 1춹 caracter칤stica
        maxX1 = max(Entradas[:, 0])       # Valor m치ximo de la 1춹 caracter칤stica
        minX2 = min(Entradas[:, 1])       # Valor m칤nimo de la 2춹 caracter칤stica
        maxX2 = max(Entradas[:, 1])       # Valor m치ximo de la 2춹 caracter칤stica
        marginX1 = (maxX1 - minX1) * 0.2  # 10% del rango
        marginX2 = (maxX2 - minX2) * 0.2  # para las dos caracter칤sticas
        x1 = np.linspace(minX1 - marginX1, maxX1 + marginX1, 1000)  # Vector de coord. X1
        x2 = np.linspace(minX2 - marginX2, maxX2 + marginX2, 1000)  # Vector de coord. X2
        X1, X2 = np.meshgrid(x1, x2)       # Generamos las matrices de coordenadas

        # Concatenaci칩n con np.c_
        Z = modelo.predict(np.c_[X1.ravel(), X2.ravel()]).reshape(X1.shape)

        
        plt.contourf(X1, X2, Z,levels = 2)

    condicion = Salidas == clase_A
    entradas_clase_A = Entradas[condicion]
    x1_A = entradas_clase_A[:,0]
    x2_A = entradas_clase_A[:,1]
    color_cero = colores[0] 
    plt.scatter(x1_A, x2_A, marker=simbolo_clase_A, s=tamano_simbolo, color=color_cero, label=etiqueta_leyenda_cero)

    condicion = Salidas == clase_B
    entradas_clase_B = Entradas[condicion]
    x1_B = entradas_clase_B[:,0]
    x2_B = entradas_clase_B[:,1]
    color_uno = colores[1]
    plt.scatter(x1_B, x2_B, marker=simbolo_clase_B, s=tamano_simbolo, color=color_uno, label=etiqueta_leyenda_uno)

    plt.title(titulo)
    plt.xlabel(titulo_eje_x)
    plt.ylabel(titulo_eje_y)
    plt.legend()
```

Ahora dibujamos los conjuntos de entrenamiento y de test y la frontera de decisi칩n. Observe que la escala de las caracter칤sticas ha cambiado respecto a la primera gr치fica en la que se mostraban las dos clases.

```python
# Figura
fig, ax = plt.subplots(figsize = (10, 7))
# Valores de los par치metros de entrada
clase_A = 1
clase_B = 2
colores = ["yellow", "magenta"]
etiqueta_leyenda_cero = 'Ent. versicolor (target 1)'
etiqueta_leyenda_uno = 'Ent. virginica (target 2)'
titulo = "Longitud p칠talo vs. Anchura p칠talo"
titulo_eje_x = "Longitud p칠talo (cm)"
titulo_eje_y = "Anchura p칠talo (cm)"
simbolo_clase_A = '+'
simbolo_clase_B = 'x'
tamano_simbolo = 50
adicion_frontera = 1
# Dibujamos el resultados correspondiente a los datos de entrenamiento
dibuja_frontera_decision(X_ent_escaladas, perceptron_monocapa, y_ent, clase_A, clase_B,
                            colores, etiqueta_leyenda_cero,
                            etiqueta_leyenda_uno, titulo, titulo_eje_x, titulo_eje_y,
                            simbolo_clase_A, simbolo_clase_B, tamano_simbolo, adicion_frontera)

# Dibujamos el resultados correspondiente a los datos de entrenamiento
etiqueta_leyenda_cero = 'Test versicolor (target 1)'
etiqueta_leyenda_uno = 'Test virginica (target 2)'
simbolo_clase_A = 'D'
simbolo_clase_B = 's'
adicion_frontera = 0
dibuja_frontera_decision(X_test_escaladas, perceptron_monocapa, y_test, clase_A, clase_B,
                            colores, etiqueta_leyenda_cero,
                            etiqueta_leyenda_uno, titulo, titulo_eje_x, titulo_eje_y,
                            simbolo_clase_A, simbolo_clase_B, tamano_simbolo, adicion_frontera)
```

**Ejercicio 1: an치lisis de la precisi칩n del perceptr칩n monocapa**

Obtenga la matriz de confusi칩n y las m칠tricas derivadas de la matriz de confusi칩n del perceptr칩n monocapa tanto para el conjunto de entrenamiento como para el conjunto de test.

쮼s un clasificador preciso? Razone la respuesta.


```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

# Definimos una funci칩n para utilizarla posteriormente
def obtencion_metricas_clasificacion(Entradas, modelo, Salidas_verdaderas):
    '''Obtenci칩n de las m칠tricas de clasificaci칩n'''
    # Salidas predichas por el modelo y las salidas verdaderas
    Salidas_predichas = modelo.predict(Entradas)
    matriz_confusion = confusion_matrix(Salidas_verdaderas, Salidas_predichas)
    print("La matriz de confusi칩n es:")
    print(matriz_confusion)

    # M칠tricas 
    exactitud = accuracy_score(Salidas_verdaderas, Salidas_predichas)
    precision = precision_score(Salidas_verdaderas, Salidas_predichas)
    sensibilidad_recall = recall_score(Salidas_verdaderas, Salidas_predichas)
    puntuacion_f1 = f1_score(Salidas_verdaderas, Salidas_predichas)

    decimales = 4
    print("La exactitud es: ")
    print(round(exactitud,decimales))
    print("La precisi칩n es:")
    print(round(precision,decimales))
    print("La sensibilidad(RVP) es:")
    print(round(sensibilidad_recall,decimales))
    print("La puntuaci칩n F1 es:")
    print(round(puntuacion_f1,decimales))


print(f"M칠tricas para el conjunto de entrenamiento:")
obtencion_metricas_clasificacion(X_ent_escaladas, perceptron_monocapa, y_ent)
print()
print(f"M칠tricas para el conjunto de test:")
obtencion_metricas_clasificacion(X_test_escaladas, perceptron_monocapa, y_test)
```

## Perceptrones multicapa. Tareas de clasificaci칩n.

**Introducci칩n**

El perceptr칩n multicapa es un tipo de red neuronal en la que existe una capa de entrada, una serie de capas ocultas y una capa de salida. La capa de entrada puede estar compuesta por varias entradas. Asimismo, la capa de salida puede estar compuesta por varias salidas. En la(s) capa(s) oculta(s) las neuronas devuelven el valor de una determinada funci칩n de activaci칩n para el valor de la entrada de la neurona. El esquema de un perceptr칩n multicapa de dos capas se muestra en la figura siguiente:



Los par치metros del perceptr칩pn son los pesos **w**  y el *bias (b)* que conectan las diferentes capas. La operaci칩n matem치tica para un perceptr칩n multicapa de dos capas es:

$$
\begin{equation}
    y_{k}=f \left( \sum_{j=1} w_{kj}^{(3)}  f \left(  \sum_{i=1} w_{ji}^{(2)} f \left(  \sum_{m=1} w_{im}^{(1)}x_{m}+ w_{m0}^{(1)} \right) + w_{i0}^{(2)} \right) + w_{k0}^{(3)} \right)
\end{equation}
$$


**Conjunto de datos**

En el entrenamiento y test de un perceptr칩n multicapa para tareas de clasificaci칩n vamos a emplear los mismos conjuntos empleados para el perceptr칩n monocapa en los apartados anteriores. As칤 podremos comparar los resultados obtenidos con el perceptr칩n monocapa y el multicapa en la tarea de clasificaci칩n.

**Construcci칩n y entrenamiento del perceptr칩n multicapa**

El perceptr칩n multicapa va a tener la siguiente estructura y par치metros de entrenamiento: 

<font color='blue'> N칰mero de capas ocultas </font>: es el par치metro `hidden_layer_sizes`. Su formato es una tupla (tuple) en el que el n칰mero de elementos marca el n칰mero de capas ocultas. El valor de cada elemento de la tupla es el n칰mero de neuronas de cada capa. Por ejemplo,  *hidden_layer_sizes=(20)* indica que hay una sola capa oculta con 20 neuronas, *hidden_layer_sizes=(20, 50)* indica que hay dos capas ocultas, la primera de ellas con 20 neuronas y la segunda con 50.

<font color='blue'> Funci칩n de activaci칩n </font>: es el par치metro `activation`. Puede tomar uno de los elementos del conjunto *{딸dentity, 딻ogistic, 땂anh, relu뗳*. La funci칩n *딸dentity* simplemente devuelve x para f(x). La funci칩n *'relu'* es la funci칩n unidad rectificada lineal. La funci칩n *'logistic'* es la funci칩n log칤stica y la funci칩n *'tanh'* es la tangente hiperb칩lica. El valor por defecto es *'relu'*.



<font color='red'> ALGORITMOS DE OPTIMIZACI칍N QUE NOS PERMITE ACTUALIZAR EL VALOR DE LOS PESOS: </font>

El <font color='blue'> m칠todo para estimar los pesos </font>: es el par치metro `solver`. Puede tomar uno de los elementos del conjunto *{딻bfgs, 땁gd, 따dam뗳*. El m칠todo *딻bfgs* es una t칠cnica de tipo quasi-Newton (algoritmo de memoria limitada Broyden뉹letcher뉺oldfarb늅hanno). El m칠todo *땁gd* es el m칠todo de descenso por gradiente estoc치stico. El m칠todo *'adam'* es el Adam estoc치stico, es el m칠todo por defecto. Como indica la ayuda de *Scikit-learn* para conjuntos de datos de entrenamiento relativamente grandes grandes el m칠todo de Adam estoc치stico ofrece buenos resultados, en cambio para conjuntos peque침os el m칠todo *딻bfgs* converge antes y ofrece mejores resultados.

<font color='blue'> N칰mero m치ximo de iteraciones </font>: es el par치metro `max_iter` y es el n칰mero m치ximo de iteraciones o epochs del m칠todo de entrenamiento. El valor por defecto es 200, NOS AYUDA A TRATAR EL SOBREAJUSTE.

<font color='blue'> Ratio o tasa de aprendizaje </font> del m칠todo de entrenamiento: es el par치metro `learning_rate`. Es la tasa de cambio en los pesos del perceptr칩n multicapa. Puede tomar uno de los valores del conjunto *{딲onstant, 딸nvscaling, 따daptive뗳*. El valor por defecto es *'constant'*. Este valor est치 fijado por otro par치metro llamado *learning_rate_init* cuyo valor por defecto es 0.001. Si se selecciona *딸nvscaling* la tasa de aprendizaje decrece gradualmente y si se selecciona *'adaptive'* la tasa decrece si se cumplen ciertas condiciones.

A continuaci칩n vamos a entrenar y validar un perceptr칩n multicapa con una 칰nica capa oculta y 5 neuronas, una funci칩n de activaci칩n log칤stica, el m칠todo de estimaci칩n de los pesos (m칠todo de entrenamiento) *딻bfgs*, un n칰mero m치ximo de iteraciones de 200 y un *learning_rate* de tipo 딸nvscaling.

```python
# Importaci칩n de la clase MLPCassifier
from sklearn.neural_network import MLPClassifier

# Instanciamos la clase MLPClassifier y generamos el perceptr칩n multicapa
# con los valores de los par치metros deseados:

# IMPORTANTE: Por ejemplo,  *hidden_layer_sizes=(20)* indica que hay una sola capa oculta con 20 neuronas,
# *hidden_layer_sizes=(20, 50)* indica que hay dos capas ocultas, la primera de ellas con 20 neuronas y la segunda con 50.

perceptron_multicapa_clasificacion = MLPClassifier(hidden_layer_sizes=(10), activation='logistic', solver='lbfgs', max_iter=200, learning_rate='invscaling')
```

Como hemos visto en la sesi칩n de estudio y an치lisis de m칠tricas, se puede evaluar el rendimiento del m칠todo de machine learning de una forma m치s completa mediante el empleo de la validaci칩n cruzada.


A continuaci칩n realizamos una operaci칩n de validaci칩n cruzada con 5 subconjuntos o folds.

```python
# Importamos la clase cross_val_score para realizar la validaci칩n cruzada
from sklearn.model_selection import cross_val_score

# El n칰mero de subconjuntos (folds) y por tanto el n칰mero de veces que se ejecuta el entrenamiento
# El valor por defecto es 5
numero_subconjuntos = 5
# Como funci칩n de utilidad tomamos la exactitud total 
funcion_utilidad = "accuracy"

# El modelo es el perceptr칩n multicapa
# EL conjunto de datos de entrada que se a dividir en diferentes subconjuntos (folds)
acierto = cross_val_score(perceptron_multicapa_clasificacion, X_ent_escaladas, y_ent,
                         scoring=funcion_utilidad, cv=numero_subconjuntos)
```

Mostramos el acierto de cada subconjunto, la media y la desviaci칩n est치ndar.

```python
# Mostramos los resultados
print("El acierto de cada subconjunto es:")
print(acierto)
# Media
decimales = 3
print("La media del acierto es:")
print(round(acierto.mean(),decimales))
# Desviaci칩n est치ndard 
print("La desviaci칩n est치ndar del acierto es:")
print(round(acierto.std(),decimales))
```

**Estimaci칩n del mejor conjunto de hiperpar치metros**

Hasta el momento hemos utilizado una configuraci칩n fija del perceptr칩n multicapa para evaluar su rendimiento. Ahora bien, dicha configuraci칩n puede ser cambiada de modo que unas configuraciones pueden ofrecer un mejor rendimiento que otras. Por ejemplo, hemos empleado un perceptr칩n multicapa de una 칰nica capa oculta, sin embargo, una estructura con m치s de una capa oculta podr칤a ofrecer mejores resultados. Asimismo, otros hiperpar치metros como el n칰mero de neuronas por capa, el propio m칠todo de entrenamiento y sus par치metros pueden ser variados para buscar una configuraci칩n mejor.  

Tal como hicimos en las sesiones anteriores vamos a emplear `GridSearchCV` para seleccionar los hiperpar치metros.

As칤, vamos a estimar la precisi칩n para un perceptr칩n multicapa de una 칰nica capa oculta con 10, 20 y 30 neuronas, con una funci칩n de activaci칩n de tres posibles tipos, log칤stico, tangente hiperb칩lica y relu y con dos m칠todos de entrenamiento, *'lbfgs'* y *'adam'*.

El n칰mero de combinaciones ser치 por tanto: 3 x 3 x 2= 18.

Vamos a tomar n칰mero de subconjuntos o *folds* igual a 5 como en el ejemplo de validaci칩n cruzada. 


Vamos a emplear la clase `Pipeline()` de *Scikit-learn* para que la estandarizaci칩n s칩lo se calcule con los datos de entrenamiento en cada iteraci칩n del algoritmo. Esta clase permite aplicar varias operaciones de forma secuencial a los datos. A continuaci칩n, importamos `GridSearchCV()` y `Pipeline` y definimos las operaciones del "pipeline".

```python
# Importamons la biblioteca para generar todas las combinaciones de par치metros de entrenamiento
from sklearn.model_selection import GridSearchCV

# Importamos la clase Pipeline para poder escalar cuando se emplea cross_val_score
from sklearn.pipeline import Pipeline

# Generamos el pipeline adecuado: emplea un estimador que es el perceptr칩n multicapa
pipeline = Pipeline([ ('transformador', scaler), ('MLP_CL', MLPClassifier())])
```

Ahora definimos la rejilla de par치metros, los par치metros de `GridSearchCV()` y creamos la variable para aplicar la operaci칩n de "grid". Observe que el modelo empleado es "pipeline" que incluye la estandarizaci칩n y al perceptr칩n multicapa de clasificaci칩n.

```python
# Los valores de los hiperpar치metros se guardan en un diccionario o en una lista de diccionarios

rejilla_hiperparametros = {
                            'MLP_CL__hidden_layer_sizes': [10, 20, 30], 
                            'MLP_CL__activation': ['logistic', 'tanh', 'relu'], 
                            'MLP_CL__solver': ['lbfgs', 'adam']
                            }

# N칰mero de folds para la validaci칩n cruzada
numero_subconjuntos = 5
# Funci칩n utilidad
funcion_utilidad = "accuracy"


# Se crean todas las configuraciones del perceptr칩n multicapa
grid_search = GridSearchCV(pipeline, rejilla_hiperparametros, cv=numero_subconjuntos,
                           scoring=funcion_utilidad, return_train_score=True)
```

Aplicamos el m칠todo de `gridsearchCV` para obtener la mejor configuraci칩n de entre las analizadas. Como la operaci칩n de estandarizaci칩n se realiza durante la operaci칩n de "grid", las caracter칤sticas se utilizan sin estandarizar.

Tambi칠n existen otras variables como la transferencia de electricidad entre estados. Vamos a considerar 칰nicamente dos variables: los periodos de tiempo correspondientes a dos d칤as consecutivos y la demanda de electricidad.

```python
# Seleccionamos el periodo y la demanda. 
# La variable x es el periodo y la variable y la demanda
caracteristica_seleccionada = "period"

#CARACTER칈STICA PERIODO:
X_reg_ini = electricidad.data[caracteristica_seleccionada]

#CARACTER칈STICA DEMANDA:
y_reg_ini = electricidad.data["nswdemand"]

# Se seleccionan los datos de dos d칤as. 
num_dias = 2
num_datos_dia = 48

# La variable dia_ini indica que el primer d칤a ser치 el siguiente al cuarto, es decir,
# el viernes. Por tanto, seleccionamos los datos del primer viernes y el primer s치bado
dia_ini = 4
ini = num_datos_dia*dia_ini
fin = ini+num_dias*num_datos_dia

rango = np.arange(ini, fin, 1) # RANGO DESDE (192, 192+96)
# NOS PERMITE COGER DE TODOS LOS DATOS LOS 칔NICOS DOS D칈AS QUE NOS INTERESAN

X_reg_ini = X_reg_ini[rango] 
y_reg_ini = y_reg_ini[rango]

lista = list(range(0, num_dias))
for indice in range(num_dias):
    rango_bucle = rango[0+indice*num_datos_dia:num_datos_dia*(indice+1)-1]
    X_reg_ini[rango_bucle] = X_reg_ini[rango_bucle] + lista[indice]
```

```python
# Programamos una funci칩n para dibujar conjuntos de puntos y que 
# opcionalmente puede dibujar la curva que genera el MLP
def dibuja_puntos(Entradas, modelo, Salidas_verdaderas, Salidas_predichas, 
                  colores, etiqueta_verdadera, etiqueta_prediccion, 
                  simbolo, tamano_simbolo, titulo, titulo_eje_x, titulo_eje_y, dibuja_curva):
    ''' Dibuja los puntos originales y las salidas generadas por el modelo de 
    machine learning '''

    color_verdaderas = colores[0]
    plt.scatter(Entradas, Salidas_verdaderas, marker=simbolo, s=tamano_simbolo, 
                color=color_verdaderas, label=etiqueta_verdadera)

    color_prediccion = colores[1]
    plt.scatter(Entradas, Salidas_predichas, marker=simbolo, 
                s=tamano_simbolo, color=color_prediccion, label=etiqueta_prediccion)

    if (dibuja_curva == 1):

        minX = min(Entradas)       # Valor m칤nimo del eje x
        maxX = max(Entradas)       # Valor m치ximo del eje y
        marginX = (maxX - minX) * 0.0  # 5 % del rango
        # Valores del eje x para dibujar la curva
        Entradas_linea = np.linspace(minX-marginX, maxX+marginX, 200)
        Salidas_linea = modelo.predict(Entradas_linea)

        plt.plot(Entradas_linea, Salidas_linea, color='green', label='Curva generada')

    plt.title(titulo)
    plt.xlabel(titulo_eje_x)
    plt.ylabel(titulo_eje_y)
    plt.legend()
```

Dibujamos los resultados. Se muestra lo siguiente:

+ Los datos del conjunto de entrenamiento con c칤rculos azules.
+ Las predicciones del conjunto de entrenamiento con c칤rculos magenta.
+ La curva que genera el modelo con una l칤nea continua en color verde.
+ Los datos del conjunto de test con diamantes negros.
+ Las predicciones del conjunto de test con diamantes verde.

```python
# Figura
fig, ax = plt.subplots(figsize = (5, 5))

colores = ["blue", "magenta"]
etiqueta_verdadera = 'Ent. verdaderas'
etiqueta_prediccion = 'Ent. predicci칩n'
titulo = "Periodo vs. demanda el칠ctrica"
titulo_eje_x = caracteristica_seleccionada
titulo_eje_y = "Demanda el칠ctrica"
simbolo = 'o'
tamano_simbolo = 50
dibuja_curva = 1
dibuja_puntos(X_reg_ent_escaladas, perceptron_multicapa_reg, y_reg_ent, y_reg_sal_ent, 
                  colores, etiqueta_verdadera, etiqueta_prediccion, 
                  simbolo, tamano_simbolo, titulo, titulo_eje_x, titulo_eje_y, dibuja_curva)


colores = ["black", "green"]
etiqueta_verdadera = 'Test. verdaderas'
etiqueta_prediccion = 'Test. predicci칩n'
simbolo = 'D'
tamano_simbolo = 50
dibuja_curva = 0
dibuja_puntos(X_reg_test_escaladas, perceptron_multicapa_reg, y_reg_test, y_reg_sal_test, 
                  colores, etiqueta_verdadera, etiqueta_prediccion, 
                  simbolo, tamano_simbolo, titulo, titulo_eje_x, titulo_eje_y, dibuja_curva)
```

```python

from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

# Definimos una funci칩n para reutilizar posteriormente
def obtencion_metricas_regresion(y, y_sal, titulo_tipo_datos):
    # Error cuadr치tico medio cometido por la t칠cnica de machine learning
    # Error de entrenamiento
    error_MSE = mean_squared_error(y, y_sal)

    mensaje = 'El error cuadr치tico medio ' + '' + titulo_tipo_datos + '' + 'es:'
    print(mensaje)
    print(round(error_MSE,4))
                                                
    # Error absoluto medio cometido por la t칠cnica de machine learning
    # Error de entrenamiento
    error_MAE = mean_absolute_error(y, y_sal)

    mensaje = 'El error absoluto medio ' + '' + titulo_tipo_datos + '' + 'es:'
    print(mensaje)
    print(round(error_MAE,4))
                                                
    # Coeficiente de determinaci칩n 
    # En el conjunto de datos de entrenamiento
    R2 = r2_score(y, y_sal)

    mensaje = 'El valor del coeficiente de determinaci칩n R2 ' + '' + titulo_tipo_datos + '' + 'es:'
    print(mensaje)
    print(round(R2,3))
                                                
    
# M칠tricas de error de regresi칩n
titulo_tipo_datos = ''
print("M칠tricas con el conjunto de entrenamiento:")
obtencion_metricas_regresion(y_reg_ent, y_reg_sal_ent, titulo_tipo_datos)

print("M칠tricas con el conjunto de test:")
obtencion_metricas_regresion(y_reg_test, y_reg_sal_test, titulo_tipo_datos)
```
