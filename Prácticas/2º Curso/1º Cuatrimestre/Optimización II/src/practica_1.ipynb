## 1.1 Existencia de solución, corcividad y regularidad

```python
from scipy.optimize import minimize_scalar
```

```python
import numpy as np
```

```python
def f(x):
  return(np.tanh(x) - 1) ** 2
```

```python
res = minimize_scalar(f)
print(f"Solución = {res.x}")
```

```python
res_2 = minimize_scalar(f, method='golden')
print(f"Solución = {res_2.x}")
```

```python
import matplotlib.pyplot as plt
```

```python
x = np.linspace(17, 19, 100)
plt.plot(x, f(x))
```

Regularizamos el problema con la regularización de Tikhonov

```python
def f_reg(x):
  alpha = 10
  return(np.tanh(x) - 1) ** 2 + alpha * x ** 2
```

```python
res = minimize_scalar(f_reg)
print(f"Solución = {res.x}")
```

```python
x = np.linspace(-1, 1, 100)
plt.plot(x, f_reg(x))
```

## 1.2 Mínimos locales, inicialización y criterio de parada

```python
def objetivo(x):
  return (x ** 2 - 1) ** 2
```

```python
from scipy.optimize import minimize
```

```python
x0 = 0 # Inicialización del algoritmo
res = minimize(objetivo, x0, method='BFGS',
               options={'gtol': 1e-6, 'disp': True})
print(f"Solución = {res.x}")
```

```python
x1 = 3 # Inicialización del algoritmo
res = minimize(objetivo, x1, method='BFGS',
               options={'gtol': 1e-6, 'disp': True})
print(f"Solución = {res.x}")
```

```python
x2 = -3 # Inicialización del algoritmo
res = minimize(objetivo, x2, method='BFGS',
               options={'gtol': 1e-6, 'disp': True})
print(f"Solución = {res.x}")
```

```python
x = np.linspace(-1.5, 1.5, 100)
plt.plot(x, objetivo(x))
```
