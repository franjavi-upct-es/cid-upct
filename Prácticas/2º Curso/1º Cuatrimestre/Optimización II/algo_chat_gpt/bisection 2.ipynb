## Método de la bisección proporcionado por ChatGPT 

```python
import numpy as np

# Define the function you want to optimize (minimize)
def f(x):
    return -(x - 2)**2 + 1  # A simple quadratic function with a minimum at x=2

# Define the derivative of the function
def df(x):
    return -2 * (x - 2)  # Derivative of the quadratic function

# Bisection method for optimization (finding the minimum)
def bisection_method_for_optimization(f, df, a, b, tol=1e-5, max_iter=100):
    """
    Perform the Bisection Method for optimization (finding the minimum).
    
    :param f: Function to be minimized.
    :param df: Derivative of the function.
    :param a: Left boundary of the interval.
    :param b: Right boundary of the interval.
    :param tol: Tolerance for convergence.
    :param max_iter: Maximum number of iterations.
    
    :return: The point x_min where the function is minimized.
    """
    if df(a) * df(b) > 0:
        print("The derivative has the same sign at both endpoints. The method may not work.")
        return None

    iter_count = 0
    while (b - a) / 2 > tol and iter_count < max_iter:
        c = (a + b) / 2  # Midpoint
        if df(c) == 0:   # If the derivative is zero, we found the minimum
            break
        #elif df(c) * df(a) < 0:  # If the sign of the derivative changes between a and c
        elif df(c)  > 0:
            b = c  # The minimum lies in [a, c]
        else:
            a = c  # The minimum lies in [c, b]
        
        iter_count += 1
    
    x_min = (a + b) / 2  # The best approximation to the minimum
    return x_min

# Parameters for the bisection method
a = 0  # Left boundary of the interval
b = 5  # Right boundary of the interval
tol = 1e-5  # Tolerance for convergence

# Apply the Bisection Method to find the minimum
x_min = bisection_method_for_optimization(f, df, a, b, tol)

# Output the result
if x_min is not None:
    print(f"Minimum found at x = {x_min}, f(x) = {f(x_min)}")
```

### Dibujamos las funciones involucradas para ver qué está pasando

```python

import matplotlib.pyplot as plt

# Define the functions
def f(x):
    return -(x - 2)**2 + 1  # The inverted quadratic function

def g(x):
    return (x - 2)**2 + 1   # The regular quadratic function

# Create an array of x values from 0 to 5
x = np.linspace(0, 5, 400)  # 400 points between 0 and 5 for a smooth curve

# Calculate the corresponding y values for both functions
y_f = f(x)
y_g = g(x)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y_f, label=r'$-(x - 2)^2 + 1$', color='red', linewidth=2)  # Plot f(x)
plt.plot(x, y_g, label=r'$(x - 2)^2 + 1$', color='blue', linewidth=2)  # Plot g(x)

# Add labels and title
plt.title('Plot of $-(x - 2)^2 + 1$ and $(x - 2)^2 + 1$', fontsize=16)
plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)

# Add a legend to differentiate between the two functions
plt.legend(fontsize=12)

# Show grid lines
plt.grid(True)

# Display the plot
plt.show()
```
