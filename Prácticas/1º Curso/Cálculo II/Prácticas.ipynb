**Actividad 57.** Estudiar la continuidad de la funci√≥n $$f(x,y)=\begin{cases}\dfrac{\sin(xy)}{x} & \text{si }x\neq0\\ y & \text{si }x=0\end{cases}$$

```python
def f(x, y):
    if x != 0:
        return sin(x*y)/x
    else:
        return y

llim = limit(f(x,y), x, 0, dir='-')
rlim = limit(f(x,y), x, 0, dir='+')

if llim == rlim == f(0,y):
    print("La funci√≥n es continua en x=0")
else:
    print("La funci√≥n no es continua en x=0")
```

**Actividad 62.** Dada la funci√≥n $$f(x,y)=x^2\tan\left(\dfrac{x^2}{x^2+y^2}\right),$$comprobar que se cumple para todo punto de su dominio la relaci√≥n $$x\dfrac{\partial f}{\partial x}(x,y)+y\dfrac{\partial f}{\partial y}(x,y)=2f(x,y).$$

**Actividad 80.** Representar gr√°ficamente las siguientes funciones de una variable.

1. $f(x)=\dfrac{1+x}{1-x^2}$ en el dominio $[-2,2]$
1. $f(x)=e^{x^2}\dfrac{1+x}{1-x^2}$ en el dominio $[-2,2]$
1. $f(x)=\sin\left(\dfrac{1+x}{1-x^2}\right)$ en el dominio $[-2,2]$
1. $f(x)=e^{x\cos x}$ en el dominio $[-5,5]$
1. $f(x)=\dfrac{e^x}{\cos x}$ en el dominio $[-\pi,\pi]$

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the functions
def f1(x):
    return (1 + x) / (1 - x**2)

def f2(x):
    return np.exp(x**2) * (1 + x) / (1 - x**2)

def f3(x):
    return np.sin((1 + x) / (1 - x**2))

def f4(x):
    return np.exp(x * np.cos(x))

def f5(x):
    return np.exp(x) / np.cos(x)

# Define the domains
domain1 = np.linspace(-2, 2, 400)
domain2 = np.linspace(-5, 5, 400)
domain3 = np.linspace(-np.pi, np.pi, 400)

# Plot the functions
plt.figure(figsize=(20, 20))

plt.subplot(3, 2, 1)
plt.plot(domain1, f1(domain1))
plt.title(r'$f(x)=\dfrac{1+x}{1-x^2}$')

plt.subplot(3, 2, 2)
plt.plot(domain1, f2(domain1))
plt.title(r'$f(x)=e^{x^2}\dfrac{1+x}{1-x^2}$')

plt.subplot(3, 2, 3)
plt.plot(domain1, f3(domain1))
plt.title(r'$f(x)=\sin\left(\dfrac{1+x}{1-x^2}\right)$')

plt.subplot(3, 2, 4)
plt.plot(domain2, f4(domain2))
plt.title(r'$f(x)=e^{x*cos(x)}$')

plt.subplot(3, 2, 5)
plt.plot(domain3, f5(domain3))
plt.title(r'$f(x)=\dfrac{e^x}{\cos(x)}$')

plt.tight_layout()
plt.show()
```

**Actividad 82.** Determinar si los siguientes puntos pertenecen a la gr√°fica de la funci√≥n que se indica:
1. El punto $(1,1,2)$ a la gr√°fica $z=x^2+y^4+2xy$
2. El punto $(1,1,0)$ a la gr√°fica $z=x^2+y^3-2xy$
3. El punto $(0,1,1)$ a la gr√°fica $z=e^x+yx$
4. El punto $(1,1,2)$ a la gr√°fica $z=x^2-y^2-5x+2y-xy$

```python
from sympy import exp

def comprobador_puntos(val1, val2, val3, f):
    x = val1
    y = val2
    z = val3
    if z == f(x,y):
        print(f"El punto ({val1},{val2},{val3}) pertenece a la gr√°fica")
    else: 
        print(f"El punto ({val1},{val2},{val3}) no pertenece a la gr√°fica")

def f1(x,y):
    return x**2+y**4+2*x*y

def f2(x,y):
    return x**2+y**3-2*x*y

def f3(x,y):
    return exp(x) + y*x

def f4(x,y):
    return x**2-y**2-5*x+2*y-x*y

comprobador_puntos(1,1,2, f1)
comprobador_puntos(1,1,0, f2)
comprobador_puntos(0,1,1, f3)
comprobador_puntos(1,1,2, f4)
```

**Actividad 84.** Dibujar conjuntamente las gr√°ficas de las funciones y planos tangentes del ejercicio anterior en un dominio apropiado. Realizar un zoom de la representaci√≥n gr√°fica anterior que muestre que el plano tangente es una aproximaci√≥n de la funci√≥n.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sympy import symbols, diff, lambdify, exp

x, y = symbols('x y')

# Definir las funciones
f1 = x**2 + y**4 + 2*x*y
f2 = x**2 + y**3 - 2*x*y
f3 = exp(x) + y*x
f4 = x**2 - y**2 - 5*x + 2*y - x*y

# Calcular las derivadas parciales
f1_dx = diff(f1, x)
f1_dy = diff(f1, y)

f2_dx = diff(f2, x)
f2_dy = diff(f2, y)

f3_dx = diff(f3, x)
f3_dy = diff(f3, y)

f4_dx = diff(f4, x)
f4_dy = diff(f4, y)

# Definir el plano tangente de las funciones
f1_tangent = lambdify((x, y), f1 + f1_dx * (x - 1) + f1_dy * (y - 1))
f2_tangent = lambdify((x, y), f2 + f2_dx * (x - 1) + f2_dy * (y - 1))
f3_tangent = lambdify((x, y), f3 + f3_dx * (x - 1) + f3_dy * (y - 1))
f4_tangent = lambdify((x, y), f4 + f4_dx * (x - 1) + f4_dy * (y - 1))

# Create a meshgrid for x, y
x_vals = np.linspace(-5, 5, 100)
y_vals = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x_vals, y_vals)

# Calculate the function values
Z1 = lambdify((x, y), f1)(X, Y)
Z2 = lambdify((x, y), f2)(X, Y)
Z3 = lambdify((x, y), f3)(X, Y)
Z4 = lambdify((x, y), f4)(X, Y)

# Calculate the tangent plane values
Z1_tangent = f1_tangent(X, Y)
Z2_tangent = f2_tangent(X, Y)
Z3_tangent = f3_tangent(X, Y)
Z4_tangent = f4_tangent(X, Y)

# Plot the functions and the tangent planes
fig = plt.figure(figsize=(12, 8))

ax1 = fig.add_subplot(221, projection='3d')
ax1.plot_surface(X, Y, Z1, alpha=0.5, rstride=100, cstride=100)
ax1.plot_surface(X, Y, Z1_tangent, color='red', alpha=0.5, rstride=100, cstride=100)

ax2 = fig.add_subplot(222, projection='3d')
ax2.plot_surface(X, Y, Z2, alpha=0.5, rstride=100, cstride=100)
ax2.plot_surface(X, Y, Z2_tangent, color='red', alpha=0.5, rstride=100, cstride=100)

ax3 = fig.add_subplot(223, projection='3d')
ax3.plot_surface(X, Y, Z3, alpha=0.5, rstride=100, cstride=100)
ax3.plot_surface(X, Y, Z3_tangent, color='red', alpha=0.5, rstride=100, cstride=100)

ax4 = fig.add_subplot(224, projection='3d')
ax4.plot_surface(X, Y, Z4, alpha=0.5, rstride=100, cstride=100)
ax4.plot_surface(X, Y, Z4_tangent, color='red', alpha=0.5, rstride=100, cstride=100)

plt.show()
```
