**Actividad 57.** Estudiar la continuidad de la función $$f(x,y)=\begin{cases}\dfrac{\sin(xy)}{x} & \text{si }x\neq0\\ y & \text{si }x=0\end{cases}$$

```python
def f(x, y):
    if x != 0:
        return sin(x*y)/x
    else:
        return y

llim = limit(f(x,y), x, 0, dir='-')
rlim = limit(f(x,y), x, 0, dir='+')

if llim == rlim == f(0,y):
    print("La función es continua en x=0")
else:
    print("La función no es continua en x=0")
```

**Actividad 62.** Dada la función $$f(x,y)=x^2\tan\left(\dfrac{x^2}{x^2+y^2}\right),$$comprobar que se cumple para todo punto de su dominio la relación $$x\dfrac{\partial f}{\partial x}(x,y)+y\dfrac{\partial f}{\partial y}(x,y)=2f(x,y).$$

```python
x, y = symbols('x y')

f = x**2*tan(x**2/(x**2+y**2))

eq = x * diff(f, x)+ y * diff(f, y) - 2*f

print("La ecuación se cumple para todo punto de su dominio:", eq.equals(0))
```

**Actividad 63.** Dada la función $$g(x,y)=\dfrac{\sqrt{x}+\sqrt{y}}{x+y},$$comprobar que se cumple para todo punto de su dominio la relación $$x\frac{\partial g}{\partial x}(x,y)+y\frac{\partial g}{\partial y}(x,y)=-\dfrac{g(x,y)}{2}.$$

```python
x, y = symbols('x y')

g = (sqrt(x) + sqrt(y)) / (x + y)
eq = x * diff(g, x) + y * diff(g, y) + g / 2

print("La ecuación se cumple para todo punto de su dominio:", eq.equals(0))
```

**Nota:** Aquí, teniendo en cuenta que $\varphi(t)$ es una función real de una variable, hay que hacer el cambio de variable $t=x+yz$ para contruir la función $f$.
