Introduce tu DNI en la forma de 1d-array.

A continuación, selecciona e imprime la siguiente información de dicho array:

1) el primer elemento
2) el último
3) El que ocupa la segunda posición (empezando desde la izquierda)
4) el que ocupa la antepenúltima 
5) los tres primeros elementos
6) los 4 últimos elementos

```python
import numpy as np
```

```python
# Completar aquí
dni = np.array([2, 6, 6, 4, 9, 1, 1, 0])
print(f"DNI = {dni}")
print(f"Primer elemento = {dni[0]}")
print(f"Último elemento = {dni[-1]}")
print(f"Segundo elemento = {dni[1]}")
print(f"Antepenúltimo elemento = {dni[-3]}")
print(f"Tres primeros elementos = {dni[0:3]}")
print(f"Cuatro último elementos = {dni[-4:]}")
# --------------------
```

Cambia la forma del 1d-array que contiene tu nombre para convertirlo en un 2d-array (matriz) de tamaño $2\times 4$, es decir dos filas y cuatro columnas. 

A continuación, selecciona e imprime los siguientes datos del nuevo 2d-array:

1) la primera fila
2) la segunda columna

```python
# Completar aquí
dni_matriz = dni.reshape(2, 4)
print(f"DNI_matriz = \n{dni_matriz}")
print(f"Primera fila = {dni_matriz[0]}")
print(f"Segunda columna = {dni_matriz[:, 1]}")

# --------------------
```

Imaginemos que el 2d-array del ejercicio anterior contiene las features $X$ y las labels $y$ de un modelo de Machine Learning. En concreto, $X$ almacena todos los datos excepto la última columna mientras que $y$ contiene los datos de la última columna.

Genera $X$ e $y$.

```python
# Completar aquí
X = dni_matriz[:, :-1]
y = dni_matriz[:, -1]
print(f"X = \n{X}")
print(f"y = {y}")
# --------------------
```

Supongamos ahora que queremos separar los datos del 2d-array en dos subconjuntos, un training dataset y un test dataset.

Training dataset será la primera fila y  test dataset será la segunda fila

```python
# Completar aquí
train_dataset = dni_matriz[0, :]
test_dataset = dni_matriz[1, :]
print(f"train dataset = {train_dataset}")
print(f"test dataset = {test_dataset}")
# --------------------
```
