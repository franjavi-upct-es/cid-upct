```python
# Importamos las librerías pandas, matplotlib y seaborn
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Indicamos que los gráficos se muestren directamente debajo de las celdas de código en el cuaderno
%matplotlib inline
```

## Extraer las tablas a partir de un documento HTML de la Wikipedia

En este documento HTML el separador de decimales es el punto y el de miles es la coma. Utilizaremos el método `read_html()` de la librería pandas para leer las tablas que hay en el fichero HTML. Con la función `len()` podemos ver el número de tablas que hemo
s leído con datos de países (no incluir la tabla de la sección *UN Forecast 2024*).

```python
# Identificamos y guardamos la URL en una variable

url = "https://en.wikipedia.org/wiki/List_of_countries_by_past_and_projected_future_population"
```

```python
# Leemos las tablas del documento HTML
tables = pd.read_html(url, index_col='Country (or dependent territory)', match='Country')
tables.pop(0)
print(f"Número de tablas encontradas: {len(tables)}")
```

## Construir una sola tabla con la información de población

Reunimos las tablas con los diferentes periodos en una sola tabla, extraemos la cabecera para inspeccionar las primeras filas de la tabla, y nos quedamos con las columnas que corresponden solamente a fechas.

```python
# Reunimos las tres tablas en una sola
df = pd.concat(tables, axis=1)

# Inspeccionamos las primeras filas de la tabla
print(df.head())

# Nos quedamos con las columnas que corresponden solamente a fechas
df = df.filter(regex=r'^\d{4}$')

# Mostramos las primeras filas de la tabla resultante
df.head(5)
```

### Distribución de la población en el histograma

La distribución de la población en el histograma es **unimodal**, ya que presenta un único pico o concentración de frecuencias en el rango más bajo de población (cercano a 0). Esto indica que la mayoría de las observaciones están agrupadas en un rango pequeño de valores, mientras que hay muy pocas observaciones en los rangos más altos.

### Implicaciones en el análisis de datos

1. **Concentración de datos**: La mayoría de los datos están concentrados en un rango muy estrecho, lo que sugiere una distribución sesgada hacia valores bajos. Esto puede ser indicativo de una gran cantidad de regiones o países con poblaciones pequeñas en comparación con unas pocas regiones con poblaciones muy grandes.
 
2. **Sesgo en la distribución**: La distribución sesgada puede dificultar el análisis, ya que los valores extremos (outliers) pueden influir desproporcionadamente en las estadísticas descriptivas, como la media.

3. **Necesidad de transformación**: Para un análisis más efectivo, puede ser necesario transformar los datos (por ejemplo, usando escalas logarítmicas) para reducir el impacto de los valores extremos y visualizar mejor las diferencias entre los grupos.

### Técnicas para segmentar y mejorar la visualización

1. **Transformación logarítmica**:
   - Aplicar una escala logarítmica a los datos de población puede comprimir los valores extremos y permitir una mejor visualización de las diferencias entre los grupos.

2. **Agrupación por categorías**:
   - Dividir los datos en categorías (por ejemplo, países con poblaciones pequeñas, medianas y grandes) puede facilitar la interpretación y el análisis.

3. **Análisis de conglomerados (clustering)**:
   - Usar técnicas como k-means o jerárquico para agrupar los datos en segmentos basados en similitudes puede ayudar a identificar patrones ocultos.

4. **Visualización alternativa**:
   - Utilizar gráficos de caja y bigotes (boxplots) para mostrar la dispersión y los valores atípicos.
   - Crear gráficos de densidad para visualizar mejor la distribución de los datos.

5. **Remuestreo o enfoque en subgrupos**:
   - Si los valores extremos dominan la visualización, se puede realizar un análisis separado para los subgrupos con poblaciones más pequeñas y más grandes.

Estas técnicas pueden ayudar a obtener una comprensión más clara de los datos y a comunicar los hallazgos de manera más efectiva.

### 1ª Mejora

```python
poblacion = df[['1950', '1980', '2010']]

print(poblacion.isna().sum())
poblacion = poblacion.dropna()
poblacion.head()
```

```python
def eliminar_outliers_iqr(df, columna, factor=2.5):
    Q1 = df[columna].quantile(0.25)
    Q3 = df[columna].quantile(0.75)
    IQR = Q3 - Q1
    umbral_inferior = Q1 - factor * IQR
    umbral_superior = Q3 + factor * IQR
    df_filtrado = df[(df[columna] >= umbral_inferior) & (df[columna] <= umbral_superior)]
    return df_filtrado

df_sin_outliers = poblacion.copy() #Copiamos el DataFrame para no modificarlo

for columna in ['1950', '1980', '2010']:
    df_sin_outliers = eliminar_outliers_iqr(df_sin_outliers, columna)

# Generamos el nuevo boxplot
plt.figure(figsize=(7, 5))
plt.boxplot(df_sin_outliers)
plt.xticks([1,2,3], [columna for columna in ['1950', '1980', '2010']])
plt.ylabel('Población (miles de millones)')
plt.title('Comparación de la distribución de la población entre 1950, 1980 y 2010')
plt.grid(True)
plt.show()
```

### 2ª Mejora

```python
import numpy as np

def transformacion_logaritmica(df, columna):
    df[columna] = np.log1p(df[columna])
    return df

df_transformado = poblacion.copy()
for año in ['1950', '1980', '2010']:
    df_transformado = transformacion_logaritmica(df_transformado, año)

# Generamos el nuevo boxplot
plt.figure(figsize=(7, 5))
plt.boxplot(df_transformado)
plt.xticks([1,2,3], [año for año in ['1950', '1980', '2010']])
plt.ylabel('Población (miles de millones)')
plt.title('Comparación de la distribución de la población entre 1950, 1980 y 2010')
plt.grid(True)
plt.show()
```

La transformación logarítmica permite normalizar y mantener a los distintos valores dentro de una escala reducida lo que facilita la visualización de los distintos patrones/métricas que se pueden interpretar en un boxplot sobre la cantidad de población en los diferentes años, comentar que la propia transformación hace que los outliers entre dentro de la escala reducida y por eso ahora no son mostrados.

### 4º Análisis de Valores Atípicos:

Los países correspodientes a estos valores atípicos son aquellos que, durante esa fecha, tuvieron un mayor crecimiento poblacional, por eso se desmarcan de la media de población que tienen todos los demás países. Como he dicho, estos valores no se deben a errores en los datos, muestran una situación real de crecimiento poblacional.

Los valores atípicos pueden tener varias implicaciones en el análisis de la población mundial:

- **Distorsión de las estadísticas descriptivas:** Los valores atípicos pueden afectar significativamente las medidas de centralidad y dispersión, como la media, la mediana y la desviación estándar.

- **Enmascaramiento de patrones:** Los valores atípicos pueden ocultar patrones o tendencias importantes en los datos.

- **Dificultad para la generalización:** Los valores atípicos pueden dificultar la generalización de los resultados del análisis a la población mundial en su conjunto.

```python
def analisis_outliers(df, columna):
    
    umbral_inferior = df[columna].quantile(0.25) - 2.5 * (df[columna].quantile(0.75) - (df[columna].quantile(0.25)))
    
    umbral_superior = df[columna].quantile(0.75) + 2.5 * (df[columna].quantile(0.75) - (df[columna].quantile(0.25)))
    
    outliers = df[(df[columna] < umbral_inferior) | (df[columna] > umbral_superior)][columna].sort_values()
    
    return outliers
```

        - La "curva de aprendizaje" puede hacerse algo compleja y a veces los errores o el método necesario para obtener la tabla específica podía volverse algo tedioso.

2. **Matplotlib**

    - **Ventajas:**

        - Debo admitir que no soy del todo crítico con esta librería ya que es la que más tiempo llevo usando y por lo tanto es a la que más acostumbrado estoy.

        - Es una librería bastante versátil y los gráficos pueden ser muy personalizables en varios aspectos.

        - Se integra muy bien con diferentes tipos de "inputs": DataFrames, ecuaciones matemáticas lineales, señales continuas y discretas, etc.

    - **Desventajas:**

        - Al igual que ocurría en el caso de Pandas, la complejidad y el tiempo de aprendizaje puede incrementarse según la complejidad de los datos que utilices y el resultado final que se espera sacar de la gráfica.
        
        - También cabe destacar que es el paquete más susceptible a errores. Por poner un ejemplo: en la representación de los *boxplots* y los diagramas de violines al usar Matplotlib había valores no válidos (como `NaN`) que no permitían visualizar los diagramas del año 2010. Esto no sucedía con Pandas o Seaborn ya que se corregían automáticamente.

3. **Seaborn:**

    - **Ventajas:**

        - Los gráficos suelen ser mucho más agradables estéticamente por defecto.
        - Integra funciones para gráficos estadístico avanzados.

    - **Desventajas:**

        - El proceso para poder ajustar los datos, al no ser compatibles con **`pd.Series`**, para los tipos de gráficas podía volverse algo lento y complicado al principio.

**Conclusión general** 

El uso combinado de las tres librerías proporciona una herramienta poderosa y una gran versatilidad para el análisis y la visualización de datos. 

Dependiendo del contexto, los requisitos del análisis y muy posiblemente la immediantez con la que se necesita el gráfico, la librería elegida puede cambiar.
