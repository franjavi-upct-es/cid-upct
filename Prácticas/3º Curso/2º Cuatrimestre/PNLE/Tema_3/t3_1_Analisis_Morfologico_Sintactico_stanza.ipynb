##3.3 Análisis sintáctico

Como hemos visto en teoría, existen dos tipos de análisis sintáctico: análisis de constituyentes (constituency parsing) y análisis de dependencias (dependency parsing)

###3.3.1 Análisis de constituyentes

A continuación, se realiza un análisis de constituyentes en español e imprimimos los árboles de constituyentes formados por las frases de ejemplo en español.

```python
import stanza
def pretty_print_tree(tree, indent=0):
    """
    Imprime un árbol de constituyentes en formato jerárquico con indentación.
    :param tree: Árbol de constituyentes (PhraseTree).
    :param indent: Nivel de indentación para las ramas.
    """
    if isinstance(tree, tuple):  # En caso de ser una hoja del árbol
        print(" " * indent + str(tree[1]))
    else:
        # Imprime la etiqueta del nodo actual
        print(" " * indent + f"({tree.label})")
        # Recorre recursivamente los hijos
        for child in tree.children:
            pretty_print_tree(child, indent + 4)


# Extrae los árboles de constituyentes
nlp = stanza.Pipeline('es', processors='tokenize,mwt, pos, constituency')
doc = nlp(sp_text)

# Extrae y guarda los árboles de constituyentes
constituency_trees = []
for i, sentence in enumerate(doc.sentences):
  tree = sentence.constituency
  constituency_trees.append(tree)

# Imprime los árboles
print("Árboles de Constituyentes:")
for i, tree in enumerate(constituency_trees):
    print(f"Oración {i}:")
    pretty_print_tree(tree)
    print("-" * 50)
```

###3.3.2 Obtenemos todos los sintagmas nominales

Con el siguiente código fuente extraemos todos los sintagmas nominales en español procesando el árbol y detectando todas las hojas pertenecientes por un árbol cuya raíz es (sn)

```python
import stanza

def obtener_hojas(sintagma):
    """
    Función recursiva para obtener las hojas más profundas de un árbol de sintagma.
    """
    if len(sintagma.children) == 0:
        return [sintagma.label]
    else:
        # Si el nodo no es una hoja, recorremos sus hijos (subárboles).
        hojas = []
        for subarbol in sintagma.children:
            hojas.extend(obtener_hojas(subarbol))
        return hojas

def extract_noun_phrases(tree):
    """
    Función para obtener los sintagmas nominales de un árbol.
    """
  # Extraemos los sintagma nominales en español
    noun_phrases = []
  # Recursión para encontrar todos los subárboles que representan sintagmas nominales
    if tree.label == "sn":  # Si el nodo es un sintagma nominal
  # Obtener las hojas más profundas del sintagma nominal
      hojas = obtener_hojas(tree)
  # Unimos las hojas para formar el sintagma completo
      sintagma_completo = " ".join(hojas)
      phrase = sintagma_completo.strip()
      noun_phrases.append(phrase)

  # Busca en los hijos del nodo actual
    for child in tree.children:
      noun_phrases.extend(extract_noun_phrases(child))
    return noun_phrases

nlp = stanza.Pipeline('es', processors='tokenize,mwt, pos, constituency')
doc = nlp(sp_text)

# Extrae e imprime los sintagmas nominales
for i, sentence in enumerate(doc.sentences, start=1):
        print(f"\nSintagmas nominales para la oración {i}:")
        noun_phrases = extract_noun_phrases(sentence.constituency)
        for np in noun_phrases:
            print(np)
        print("-" * 50)
```

```python
#Modificamos la función para que se pueda meter el idioma inglés "en" o español "es"
def extract_noun_phrases(tree, idioma):
    """
    Función para obtener los sintagmas nominales de un árbol.
    """
    if(idioma == 'es'):
      label = "sn"
    if(idioma == 'en'):
      label = "NP"
  # Extraemos los sintagma nominales en español
    noun_phrases = []
  # Recursión para encontrar todos los subárboles que representan sintagmas nominales
    if tree.label == label:  # Si el nodo es un sintagma nominal
  # Obtener las hojas más profundas del sintagma nominal
      hojas = obtener_hojas(tree)
  # Unimos las hojas para formar el sintagma completo
      sintagma_completo = " ".join(hojas)
      phrase = sintagma_completo.strip()
      noun_phrases.append(phrase)

  # Busca en los hijos del nodo actual
    for child in tree.children:
      noun_phrases.extend(extract_noun_phrases(child,idioma))
    return noun_phrases

nlp = stanza.Pipeline('en', processors='tokenize,mwt, pos, constituency')
doc = nlp(en_text)

# Extrae e imprime los sintagmas nominales
for i, sentence in enumerate(doc.sentences, start=1):
        print(f"\nSintagmas nominales para la oración {i}:")
        noun_phrases = extract_noun_phrases(sentence.constituency, 'en')
        for np in noun_phrases:
            print(np)
        print("-" * 50)
```

###3.3.5 Análisis de dependencias
Ahora incluimos en el Pipeline que se haga un análisis de dependencias (*depparse*)

Obtenemos entonces por cada frase la raiz principial (*root*) y su sujeto (*nsubj*).

(Opcional) Mostrar también sus complementos.

Los valores que toma el atributo *deprel* están detallados en esta referencia https://universaldependencies.org/u/dep/

```python
pipelineStanza = stanza.Pipeline(lang='es', processors='tokenize, mwt, pos, lemma, depparse')
stanzaDoc = pipelineStanza(sp_text)

for sentence in stanzaDoc.sentences:
    print(sentence)
    frase = ""
    raiz = ""
    sujeto = ""
    obj= ""
    print("Frase")
    for word in sentence.words:
        frase = frase + word.text + "(" + word.lemma +")_" + word.upos + " "
        if(word.deprel == "root"):
            raiz = word.text + "(" + word.lemma +")_" + word.upos
        if(word.deprel == "nsubj"):
            sujeto = word.text + "(" + word.lemma +")_" + word.upos
        if(word.deprel == "obj"):
            obj = obj + word.text + "(" + word.lemma +")_" + word.upos +" "
    print(frase)
    print("Sujeto: "+ sujeto +"\t Raiz (verbo):"+ raiz +"\t Complemento directo:" + obj)
    print("-----------------------------------------------------------")
```

```python
def obtener_sintagma_completo(sentence, word):
  ids= []
  sintagma = [word]  # Incluye el núcleo del sintagma
  ids.append(word.id)
  while(ids): # mientras haya elementos vamos buscando si hay palabras que referencian
  # como head a algún id
    for w in sentence.words:
      if w.head in ids:
        if(w.id not in ids):
          ids.append(w.id)
          sintagma.append(w)
        # Sacamos el último elemento
    ids.pop(0)
  # Combina los elementos del sintagma completo y lo devuelve
  return " ".join(w.text for w in sorted(sintagma, key=lambda x: x.id))


pipelineStanza = stanza.Pipeline(lang='es', processors='tokenize, mwt, pos, lemma, depparse')
stanzaDoc = pipelineStanza(sp_text)

for sentence in stanzaDoc.sentences:
    frase = ""
    raiz = ""
    sujeto = ""
    obj= ""
    print("Frase")
    for word in sentence.words:
        frase = frase + word.text + "(" + word.lemma +")_" + word.upos + " "
        if(word.deprel == "root"):
            raiz = word.text
        if(word.deprel == "nsubj"):
            sujeto = obtener_sintagma_completo(sentence,word)
        if(word.deprel == "obj"):
            obj = obtener_sintagma_completo(sentence,word)
    print(frase)
    print("Sujeto: "+ sujeto +"\t Raiz (verbo):"+ raiz +"\t Complemento directo:" + obj)
    print("-----------------------------------------------------------")
```
