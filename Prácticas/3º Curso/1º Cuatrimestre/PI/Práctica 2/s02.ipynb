## Revisión de Conceptos

### *Formatos de Imagen*

Los archivos de imagen pueden adoptar diversos formatos, cada uno con características específicas que los hacen adecuados para distintos usos. A continuación se presenta una breve descripción de los tipos de archivos de imagen más comunes:

1. **[BMP](https://en.wikipedia.org/wiki/BMP_file_format) (Bitmap)**: Este formato no utiliza compresión, lo que garantiza la máxima calidad de imagen sin pérdida. Puede manejar desde 1 hasta 32 bits por píxel, lo que permite una amplia gama de colores, haciéndolo ideal para almacenar imágenes de alta fidelidad en aplicaciones donde la calidad de imagen es prioritaria.

2. **[PNG](https://en.wikipedia.org/wiki/PNG) (Portable Network Graphics)**: PNG es famoso por su compresión sin pérdida y soporte de transparencia (canal alfa), lo que lo hace perfecto para imágenes detalladas como gráficos web. Es especialmente útil para imágenes que requieren detalles precisos y colores ricos sin comprometer la calidad.

3. **[JPEG](https://en.wikipedia.org/wiki/JPEG) (Joint Photographic Experts Group)**: JPEG utiliza una compresión con pérdida que reduce significativamente el tamaño del archivo, lo cual es perfecto para fotografías y otros contenidos visuales donde el tamaño de archivo puede ser un factor limitante. La compresión puede ajustarse para equilibrar la calidad y el tamaño del archivo según las necesidades del usuario.

4. **[GIF](https://en.wikipedia.org/wiki/GIF) (Graphics Interchange Format)**: GIF es ideal para animaciones simples y gráficos con hasta 256 colores. Es popular para gráficos en movimiento y banners en sitios web debido a su capacidad de compresión sin pérdida para paletas de colores limitadas.

5. **[PPM](https://en.wikipedia.org/wiki/Netpbm) y [PNM](https://en.wikipedia.org/wiki/Netpbm) (Netpbm format)**: Estos formatos se caracterizan por su simplicidad, lo que los hace ideales para tareas académicas y de procesamiento básico de imágenes. Aunque no soportan compresión ni transparencia, su uso sigue siendo práctico en contextos educativos y de desarrollo.

6. **[TIFF](https://en.wikipedia.org/wiki/TIFF) (Tagged Image File Format)**: TIFF es capaz de almacenar imágenes de alta calidad con soporte para múltiples capas y canales de color. Es ampliamente utilizado en la edición profesional de imágenes y la impresión debido a su capacidad para manejar compresión sin pérdida y su extensibilidad que permite almacenar diversos tipos de datos de imagen.

### *Formatos de Color*

Las imágenes digitales no solo se clasifican por sus formatos de archivo, sino también por la estructura de datos interna que utilizan. Esta estructura se define principalmente por la cantidad de canales que contienen las imágenes y la información específica que cada canal representa. De esta forma, podemos distinguir principalmente dos categorías:

1. **Imágenes de Un Solo Canal (Monocromáticas)**: Cada píxel en estas imágenes está representado por un único valor. Dentro de esta categoría podemos identificar:
    - **Imágenes Binarias**: Cada píxel se representa con un solo bit, que puede ser 0 (negro) o 1 (blanco).
    - **Imágenes en Escala de Grises**: Cada píxel se representa con un valor de bit más amplio, típicamente 8 bits, lo que permite valores de 0 a 255, abarcando gradientes de gris desde negro hasta blanco.

2. **Imágenes de Múltiples Canales**: Cada píxel en estas imágenes se representa mediante una combinación de valores, siendo algunos de los más habituales:
    - **Imágenes RGB**: Cada píxel se compone de tres valores (R, G, B), que representan las intensidades de los colores rojo, verde y azul, respectivamente.
    - **Imágenes HSV**: Se caracterizan por una tupla (H, S, V) para cada píxel, indicando el matiz (hue), la saturación (cantidad de color mezclado con blanco), y el valor (brillo del color mezclado con negro).
    - **Imágenes RGBA**: Similar a RGB, pero con un cuarto canal alfa ($\alpha$) que añade la transparencia a cada píxel, permitiendo efectos de superposición y transparencias en composiciones de imagen.

## Archivos de imagen en Python

Las bibliotecas de procesamiento de imágenes en Python identifican automáticamente el formato de una imagen a través de la extensión del archivo, extrayendo toda la información de la propia imagen, como el tamaño, el tipo y el formato de los datos de la imagen. Pese a que existe un alto grado de compatibilidad, no todas las librerías soportan todos los formatos de archivo de imagen. Por tanto, es esencial conocer qué formatos puede manejar cada biblioteca para elegir la herramienta más adecuada en cada momento. La siguiente tabla resume la compatibilidad de los formatos más comunes con las bibliotecas PIL (Pillow), scikit-image, Matplotlib y OpenCV de Python.

| **Formato de Imagen** | **PIL (Pillow)** | **scikit-image** | **Matplotlib** | **OpenCV** |
|-----------------------|------------------|------------------|----------------|------------|
| BMP                   | Sí               | Sí               | Sí             | Sí         |
| PNG                   | Sí               | Sí               | Sí             | Sí         |
| JPEG                  | Sí               | Sí               | Sí             | Sí         |
| GIF                   | Sí               | Sí               | Sí             | No         |
| TIFF                  | Sí               | Sí               | Sí             | Sí         |
| PPM                   | Sí               | Sí               | Sí             | Sí         |
| PNM                   | Sí               | Sí               | No             | No         |


#### Observaciones:
- **PIL (Pillow)**: Es muy versátil en cuanto a formatos de archivo compatibles, siendo una de las opciones más aconsejables.
- **scikit-image**: Utiliza principalmente el backend de PIL para cargar imágenes, por lo que soporta muchos de los mismos formatos.
- **Matplotlib**: Debido a que está orientada a la visualización, Matplotlib permite cargar y mostrar una amplia variedad de tipos de imágenes.
- **OpenCV**: Excelente para tareas de procesamiento de imágenes en tiempo real, pero no soporta nativamente GIF ni PNM.

### *Conversión entre formatos de imagen*

Usando PIL, podemos leer una imagen en un formato de archivo concreto y guardarla en otro; por ejemplo, de `PNG` a `JPG`, como se muestra a continuación.

```python
from PIL import Image

im = Image.open("./images/parrot.png")
print(im.mode)

im.save("parrot.jpg")
```

No obstante, si la imagen cargada se encuentra en el modo de color `RGBA`, necesitamos primero convertirla a `RGB`, ya que el formato de archivo `JPG` no es compatible con el canal alfa de transparencia. El siguiente código muestra cómo convertir y guardar la imagen en este caso.

```python
im = Image.open("./images/hill.png")
print(im.mode)

im.convert('RGB').save("hill.jpg")
```

### *Conversión entre estructuras*

Como ya sabemos, PIL utiliza el objeto `Image` para almacenar una imagen, mientras que otras librerías, como scikit-image, Matplolib y OpenCV, utilizan la estructura de datos `ndarray` de Numpy para almacenar los datos de la imagen. El siguiente bloque de código muestra cómo convertir un objeto de imagen de PIL a un array de Numpy para ser manipulado por otras librerías.

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from skimage.io import imshow, show

im = Image.open('./images/flowers.png')
im = np.array(im)

imshow(im); plt.axis('off'); show()
```

Para realizar la conversión inversa, es decir, de un array de Numpy a un objeto de imagen de PIL, basta con invocar a la función `fromarray()`.

```python
from skimage.io import imread

im = imread('./images/flowers.png') 
print(f"{type(im)}")

im = Image.fromarray(im) 
print(f"{type(im)}")
```

### *Conversiones entre modos de color*

Es posible convertir una imagen RGB a escala de grises directamente durante el proceso de lectura de la misma. El código que sigue muestra cómo realizar esta conversión utilizando scikit-image.

```python
from skimage.io import imread

im = imread("./images/parrot.png", as_gray=True)
print(im.shape)
```

Es importante destacar que al convertir imágenes a color a escala de grises podemos perder información visual crucial. Para ilustrar esto, presentamos un ejemplo utilizando placas de Ishihara, que son empleadas comúnmente para detectar el daltonismo. El código aplicado utiliza la función `rgb2gray()` del módulo `color` para realizar la conversión. Las imágenes resultantes, una en color y otra en escala de grises, se muestran una al lado de la otra para facilitar la comparación. Como se puede observar en la figura adjunta, el número 8 en la placa se vuelve casi invisible en la versión en escala de grises.

```python
import matplotlib.pyplot as plt
from skimage import color

im = imread("./images/Ishihara.png")
im_g = color.rgb2gray(im)
plt.subplot(121); plt.imshow(im); plt.axis('off')
plt.subplot(122); plt.imshow(im_g, cmap='gray'); plt.axis('off')
plt.show()
```

Es posible realizar conversiones entre diferentes espacios de color utilizando las funciones disponibles en scikit-image. Por ejemplo, el siguiente código muestra cómo transformar una imagen desde el espacio de color `RGB` al espacio de color `HSV`.

```python
im = imread("./images/parrot.png")
im_hsv = color.rgb2hsv(im)

plt.figure(figsize=(10,8))
plt.subplot(221); plt.imshow(im_hsv); plt.title('HSV', size=20); plt.axis('off')
plt.subplot(222); plt.imshow(im_hsv[...,0]); plt.title('H', size=20); plt.axis('off')
plt.subplot(223); plt.imshow(im_hsv[...,1]); plt.title('S', size=20); plt.axis('off')
plt.subplot(224); plt.imshow(im_hsv[...,2]); plt.title('V', size=20); plt.axis('off')
plt.show()
```

## Manipulaciones básicas en Python

### *Máscara circular*

El siguiente fragmento de código ilustra cómo utilizar operaciones de corte y enmascaramiento con arrays de Numpy para generar una máscara circular sobre una imagen. Este ejemplo carga una imagen utilizando Matplotlib, calcula sus dimensiones y crea una matriz de coordenadas. La máscara circular se define centrada en la imagen, y los píxeles que caen fuera de esta área se ponen a cero, efectivamente ocultándolos. Finalmente, la imagen modificada se visualiza sin ejes para destacar el efecto de la máscara.


```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np

lena = mpimg.imread("./images/lena.jpg")

lx, ly, _ = lena.shape
X, Y = np.ogrid[0:lx, 0:ly]

mask = (X - lx / 2) ** 2 + (Y - ly / 2) ** 2 > lx * ly / 4
lena_masked = lena.copy()
lena_masked[mask,:] = 0

plt.figure(figsize=(5,5))
plt.imshow(lena_masked); plt.axis('off'); plt.show()
```

### *Disolución de imágenes*

El siguiente código realiza una transición visual entre dos imágenes utilizando Numpy. Comenzamos con el rostro de Lionel Messi y progresivamente cambiamos al de Cristiano Ronaldo. Esta transición se logra mediante una combinación lineal de los dos arrays de imágenes, empleando la siguiente fórmula, donde $\alpha$ es el factor de disolución y varía entre 0 y 1.

$$
(1 - \alpha) \cdot \text{image}_1 + \alpha \cdot \text{image}_2
$$

Como se puede observar, la imagen de Messi se desvanece gradualmente a medida que $\alpha$ aumenta, mientras que la de Ronaldo se vuelve más prominente. El código visualiza este efecto de disolución en una serie de 20 pasos, mostrando el cambio progresivo en una cuadrícula.

```python
im1 = mpimg.imread("./images/messi.jpg") / 255
im2 = mpimg.imread("./images/ronaldo.jpg") / 255

i = 1
plt.figure(figsize=(18,15))
for alpha in np.linspace(0,1,20):
    plt.subplot(4,5,i)
    plt.imshow((1-alpha)*im1 + alpha*im2)
    plt.axis('off')
    i += 1
plt.subplots_adjust(wspace=0.05, hspace=0.05)
plt.show()
```

### *Recorte y pegado*

Aunque siempre es posible manipular las imágenes como arreglos de NumPy, PIL ofrece métodos específicos que simplifican algunos procesos. El siguiente código utiliza el método `crop()` para recortar un área rectangular de una imagen, definida por las coordenadas (175, 75) y (320, 200). Seguidamente, crea una nueva imagen de color gris mediante `new()` y pega la región recortada en las mismas coordenadas originales con el método `paste()`.

```python
from PIL import Image
from IPython.display import display

im = Image.open("./images/parrot.png")
print(im.width, im.height, im.mode, im.format)

im_c = im.crop((175,75,320,200))

im_p = Image.new(mode='RGB', size=(im.width, im.height), color=(60,60,60))
im_p.paste(im_c, (175,75))

display(im); display(im_c); display(im_p)
```

### *Rescalado de imagen*

El rescalado es una técnica común en el procesamiento de imágenes que se utiliza para ajustar el tamaño de una imagen, ya sea aumentándolo o reduciéndolo. Mediante la biblioteca PIL en Python, podemos cambiar fácilmente las dimensiones de una imagen utilizando el método `resize()`. Este método permite especificar nuevas dimensiones y seleccionar un algoritmo de interpolación, como el Bicúbico (`BICUBIC`), que ayuda a mantener la suavidad de las líneas y la calidad general al rescalar. En el código proporcionado, se carga una imagen de un reloj y se muestra su tamaño original. Luego, se genera y visualiza una versión aumentada al doble de su tamaño original y otra reducida a la mitad de su tamaño inicial.

```python
from PIL import Image
from IPython.display import display

im = Image.open("./images/clock.jpg")
print(im.width, im.height)
display(im)

im_large = im.resize((im.width*2, im.height*2), Image.BICUBIC)
display(im_large)

im_small = im.resize((im.width//2, im.height//2), Image.BICUBIC)
display(im_small)
```

### *Reflejos y rotaciones*

La librería PIL ofrece también funciones muy útiles para modificar directamente la orientación y el ángulo de las imágenes. En el siguiente bloque de código, se aplica una operación de horizontal utilizando el método `transpose()` con el argumento `FLIP_LEFT_RIGHT`. Posteriormente, la imagen se rota 45 grados usando el método `rotate()`.


```python
from PIL import Image
from IPython.display import display

im = Image.open("./images/parrot.png")

im_tr = im.transpose(Image.FLIP_LEFT_RIGHT)
display(im_tr)

im_45 = im.rotate(45) 
display(im_45)
```

Tal y como puedes observar, al rotar una imagen partes de ella pueden quedar fuera del marco visible y parecer cortadas. Para evitar este efecto, puedes usar el parámetro `expand` establecido en True. Este ajuste expande automáticamente el lienzo para acomodar la imagen completa. Adicionalmente, puedes definir el color de relleno para los espacios vacíos que se crean durante la rotación utilizando el parámetro `fillcolor`. En el siguiente ejemplo, se utiliza el color transparente (0,0,0,0) para mantener la consistencia visual de la imagen original.

```python
im_45_rgba = im.convert('RGBA')
im_45_rgba = im_45_rgba.rotate(45, expand=True, fillcolor=(0,0,0,0))
display(im_45_rgba)
```

A continuación, encuentra un umbral adecuado para identificar los píxeles oscuros que se considerarán fondo y utiliza este umbral para hacer transparentes dichos píxeles. Finalmente, guarda la imagen procesada como un archivo PNG y muestra por pantalla tanto la imagen original como la modificada para comprobar el resultado.

### *Ejercicio 2. Simulación de Daltonismo*

Desarrolla un programa en Python que utilice las librerías PIL y Numpy para simular un tipo de daltonismo que afecta a la capacidad de distinguir los colores rojo y verde. Comienza cargando la imagen `./images/Ishihara.png` y ajusta los canales rojo y verde para reflejar cómo se percibirían con este tipo de condición visual. Posteriormente, muestra por pantalla la imagen original y la imagen procesada, guardando esta última en disco. Finalmente, selecciona otra imagen de Internet y repite el proceso para confirmar la consistencia del efecto simulado.

<div style="text-align: center; margin-bottom: 20px;">
    <h3>Resultado esperado</h3>
</div>
<div style="display:flex; justify-content:center; align-items:center;">
    <img src="images/Ishihara.png" style="width:20%; height:auto; margin-right: 50px;">
    <img src="images/ej2_Ishihara.png" style="width:20%; height:auto; margin-right: 50px;">
</div>
<br><br>

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# Cargamos la imagen
img = Image.open("./images/Ishihara.png").convert("RGB")
arr = np.array(img).astype(np.float32)

# Separamos los canales
r, g, b = arr[..., 0], arr[..., 1], arr[..., 2]

# Hacemos que el rojo y el verde se parezcan para simular el daltonismo
mezcla = (r + g) / 2.0
fuerza = 0.95 # 0 = nada, 1 = se confunden del todo

r_sim = (1 - fuerza) * r + fuerza * mezcla
g_sim = (1 - fuerza) * g + fuerza * mezcla
b_sim = b # El azul no se toca

# Reconstruimos la imagen
out_arr = np.stack([r_sim, g_sim, b_sim], axis=-1)
out_arr = np.clip(out_arr, 0, 255).astype(np.uint8)
img_sim = Image.fromarray(out_arr, "RGB")

# Resultados
img_sim.save("Ishihara_daltonica.png")

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(img)
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(img_sim)
plt.axis("off")
plt.show()
```

Las fotografías Polaroid, especialmente las del modelo 600, son conocidas por su diseño único y dimensiones específicas. Una foto Polaroid típica mide aproximadamente 3.5 x 4.25 pulgadas, con un área de imagen visible de 3.1 x 3.1 pulgadas. Esta área está rodeada por un margen blanco de 0.2 pulgadas en los lados superior, izquierdo y derecho, y un margen más ancho de 0.95 pulgadas en la parte inferior. En este ejercicio, vas a crear un script en Python que, utilizando las bibliotecas PIL y Numpy, simule el efecto característico de las fotografías Polaroid sobre una imagen dada. Debes asegurarte de que el tamaño final de la imagen procesada sea de 350 x 425 píxeles, realizando los ajustes de escala necesarios para mantener las proporciones de la zona visible y los márgenes típicos de una Polaroid.

```python
from PIL import Image, ImageOps
import matplotlib.pyplot as plt

# Parámetros Polaroid (px) equivalentes a 3.5" x 4.25" a ~100 px/pulgada
W_TOTAL, H_TOTAL = 350, 425
MARGIN_L = 20   # 0.2"
MARGIN_R = 20   # 0.2"
MARGIN_T = 20   # 0.2"
MARGIN_B = 95   # 0.95"
W_IMG = 310     # 3.1"
H_IMG = 310     # 3.1"

# 1) Cargamos la imagen original
img = Image.open("./images/lena.jpg").convert("RGB")

# 2) Ajustamos, manteniendo la proporción para que encaje en 310x310
img_sq = ImageOps.fit(img, (W_IMG, H_IMG), method=Image.Resampling.LANCZOS)

# 3) Creamos el lienzo blanco 350x425 (marco Polaroid)
canvas = Image.new("RGB", (W_TOTAL, H_TOTAL), color=(255,255,255))

# 4) Pegamos la imagen cuadrada en la posición correcta
canvas.paste(img_sq, (MARGIN_L, MARGIN_T))

# Resultados
canvas.save("lena_polaroid.jpg")

plt.figure()
plt.imshow(canvas)
plt.axis("off")
plt.show()
```

### *Ejercicio 4. Tablón de fotos*

En esta actividad, vas a desarrollar una extensión del ejercicio anterior en el que creaste un efecto Polaroid para la imagen de Lena. Utilizando las bibliotecas PIL y Numpy, tu tarea ahora será tomar la imagen de Lena ya procesada, añadirle el texto "Lena" en el margen inferior como si estuviera firmada, y dibujar un círculo gris en el centro del margen superior para simular una chincheta. Posteriormente, rotarás la imagen aleatoriamente entre -25 y +25 grados para darle un aspecto de foto colocada al azar en un tablón. Finalmente, integrarás esta imagen en una imagen de fondo más grande que simula un tablero de corcho (disponible en `./images/tablero.png`), colocando la imagen de Lena en una posición centrada. El resultado final deberá ser mostrado en pantalla y guardado en el disco.

```python
from PIL import Image, ImageDraw, ImageFont, ImageOps
import matplotlib.pyplot as plt
import numpy as np
import random

# Rutas
IMG_LENA = "lena_polaroid.jpg"
IMG_TABLERO = "./images/tablero.png"
OUT_IMG = "lena_polaroid_tablero.png"

# Cargamos las imágenes
polaroid = Image.open(IMG_LENA).convert("RGBA")
tablero = Image.open(IMG_TABLERO).convert("RGBA")

W_TOTAL, H_TOTAL = polaroid.size
MARGIN_L = 20
MARGIN_T = 20
MARGIN_B = 95

# Escribimos "Lena" en la parte inferior del polaroid
# Intentamos una fuente TTF; si no, caemos a la fuente por defecto.
draw = ImageDraw.Draw(polaroid)
try:
    font = ImageFont.truetype("./images/arial.ttf", 20)
except:
    font = ImageFont.load_default()

texto = "Lena"
# Posición centrada horizontalmente; vertical dentro de la imagen inferior
tw, th = draw.textbbox((0, 0), texto, font=font)[2:]
x_text = (W_TOTAL - tw) // 2
y_text = H_TOTAL - MARGIN_B + (MARGIN_B - th)//2
draw.text((x_text, y_text), texto, fill=(20,20,20,255), font=font)

# Dubujamos la chincheta
pin_radius = 8
cx = W_TOTAL // 2
cy = MARGIN_T // 2
bbox = [cx - pin_radius, cy - pin_radius, cx + pin_radius, cy + pin_radius]
draw.ellipse(bbox, fill=(180, 180, 180, 255), outline=(180, 180, 180, 255), width=2)

# Rotación aleatoria entre -25 y +25 grados
angle = random.uniform(-25, 25)
polaroid_rot = polaroid.rotate(angle, resample=Image.Resampling.BICUBIC, expand=True, fillcolor=(0,0,0,0))

# Los componemos sobre el tablero de corcho
tab_w, tab_h = tablero.size
pw, ph = polaroid_rot.size
x = (tab_w - pw) // 2
y = (tab_h - ph) // 2

composite = tablero.copy()
composite.alpha_composite(polaroid_rot, (x, y))

# Resultados
composite.save(OUT_IMG)

plt.figure(figsize=(10,10))
plt.imshow(composite)
plt.axis("off")
plt.show()
```
