```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.filters import threshold_otsu
from skimage.morphology import binary_opening, binary_closing, disk
from skimage.measure import label, regionprops
from skimage.io import imread
from skimage.color import rgb2gray

dice = imread('./images/imagen_entrada.png')
dice_gray = rgb2gray(dice)

# 1) Umbralizaci√≥n (Otsu)
th = threshold_otsu(dice_gray)
# como los puntos son oscuros:
binary = dice_gray < th     # True = puntos

# 2) Limpieza morfol√≥gica
selem = disk(2)  # tama√±o ajustable seg√∫n los puntos
binary_clean = binary_opening(binary, selem)     # eliminar ruido peque√±o
binary_clean = binary_closing(binary_clean, selem)  # cerrar peque√±os huecos

# 3) Etiquetado de componentes conectados
labels = label(binary_clean)

# 4) Visualizaci√≥n r√°pida
fig, axes = plt.subplots(1, 3, figsize=(10, 4))
axes[0].imshow(dice_gray, cmap='gray'); axes[0].set_title('Original'); axes[0].axis('off')
axes[1].imshow(binary_clean, cmap='gray'); axes[1].set_title('Binaria limpia'); axes[1].axis('off')
axes[2].imshow(labels, cmap='nipy_spectral'); axes[2].set_title('Componentes'); axes[2].axis('off')
plt.tight_layout(); plt.show()
```

```python
import numpy as np
from skimage.morphology import binary_erosion, binary_dilation, disk
from skimage.measure import label

mask = binary  # m√°scara: posibles puntos

# 1) Semilla por erosi√≥n: reduce los puntos
selem_seed = disk(2)
seed = binary_erosion(mask, selem_seed)

# 2) Crecimiento por dilataci√≥n limitada por la m√°scara (reconstrucci√≥n morfol√≥gica manual)
grow = seed.copy()
selem_grow = disk(1)

while True:
    new = binary_dilation(grow, selem_grow) & mask  # dilata pero no sale de la m√°scara
    if np.array_equal(new, grow):
        break
    grow = new

pips_seed_method = grow

labels_seed = label(pips_seed_method)

# Visualizaci√≥n
fig, axes = plt.subplots(1, 3, figsize=(10, 4))
axes[0].imshow(mask, cmap='gray'); axes[0].set_title('M√°scara inicial'); axes[0].axis('off')
axes[1].imshow(seed, cmap='gray'); axes[1].set_title('Semilla (erosionada)'); axes[1].axis('off')
axes[2].imshow(labels_seed, cmap='nipy_spectral'); axes[2].set_title('Resultado crecimiento'); axes[2].axis('off')
plt.tight_layout(); plt.show()
```

```python
from skimage.filters import sobel
from skimage.segmentation import watershed
from scipy import ndimage as ndi

# 1) Mapa de elevaci√≥n a partir del gradiente (Sobel)
elevation_map = sobel(dice_gray)

# 2) Marcadores: fondo y posible objeto seg√∫n intensidad
markers = np.zeros_like(dice_gray, dtype=int)
markers[dice_gray > 0.8] = 1   # fondo (muy blanco)
markers[dice_gray < 0.4] = 2   # posibles puntos (oscuros)

# 3) Watershed
segmentation = watershed(elevation_map, markers)

# segmentaci√≥n == 2 ser√≠an las regiones de "puntos"
pips_ws = segmentation == 2

# Limpieza opcional con morfolog√≠a
pips_ws_clean = binary_opening(pips_ws, disk(1))

labels_ws = label(pips_ws_clean)

fig, axes = plt.subplots(1, 3, figsize=(10, 4))
axes[0].imshow(dice_gray, cmap='gray'); axes[0].set_title('Original'); axes[0].axis('off')
axes[1].imshow(elevation_map, cmap='gray'); axes[1].set_title('Mapa Sobel'); axes[1].axis('off')
axes[2].imshow(labels_ws, cmap='nipy_spectral'); axes[2].set_title('Puntos (WS)'); axes[2].axis('off')
plt.tight_layout(); plt.show()
```

```python
from skimage.feature import canny
from skimage.morphology import dilation, footprint_rectangle

edges = canny(labels_ws, sigma=1.5, low_threshold=0.05, high_threshold=0.15)

plt.imshow(edges)

# Dilatar los bordes para hacer m√°s visibles las embarcaciones
edges_dilated = dilation(edges, footprint_rectangle((2, 2)))

# Crear imagen con bordes resaltados en rojo
image_with_edges = dice.copy()
# Convertir bordes a m√°scara RGB (rojo)
red_mask = np.zeros_like(dice)
red_mask[edges_dilated > 0] = [255, 0, 0] # Rojo

image_with_edges = np.where(red_mask > 0, red_mask, dice)

plt.imshow(image_with_edges)
```
