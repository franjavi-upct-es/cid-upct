# Cuestin 2 - # Control Difuso de Temperatura en un Horno Industrial (7 puntos)

Se desea implementar un sistema de control difuso para regular la **temperatura en un horno industrial**. Para ello, se emplearn como **variables de entrada**:

   - *Si la temperatura es media y la carga es baja  potencia alta*

```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
from skfuzzy import control as ctrl
```

```python
# Definimos los universos de cada una de las variables
temperatura = np.arange(0, 301, 1)
carga = np.arange(0, 101, 1)
potencia = np.arange(0, 101, 1)
```

```python
# Funciones trapezoidales de las variables
temperatura_alta = fuzz.trapmf(temperatura, [150, 200, 250, 301])
temperatura_media = fuzz.trapmf(temperatura, [80, 110, 140, 170])
temperatura_baja = fuzz.trapmf(temperatura, [-1, -1, 50, 100])

carga_alta = fuzz.trapmf(carga, [60, 70, 100, 101])
carga_media = fuzz.trapmf(carga, [30, 40, 60, 70])
carga_baja = fuzz.trapmf(carga, [-1, -1, 20, 40])

potencia_alta = fuzz.trapmf(potencia, [60, 70, 100, 101])
potencia_media = fuzz.trapmf(potencia, [30, 40, 60, 70])
potencia_baja = fuzz.trapmf(potencia, [-1, -1, 20, 40])
```

```python
# Definimos las variables de entrada (antecedentes) y de salida (consecuente)
temperatura_input = ctrl.Antecedent(temperatura, 'temperatura')
carga_input = ctrl.Antecedent(carga, 'carga')

potencia_input = ctrl.Consequent(potencia, 'potencia')
```

```python
# Asignamos las funciones de pertenencia a cada etiqueta lingistica
temperatura_input['alta'] = temperatura_alta
temperatura_input['media'] = temperatura_media
temperatura_input['baja'] = temperatura_baja

carga_input['alta'] = carga_alta
carga_input['media'] = carga_media
carga_input['baja'] = carga_baja

potencia_input['alta'] = potencia_alta 
potencia_input['media'] = potencia_media
potencia_input['baja'] = potencia_baja
```

```python
# Visualizar las funciones de pertenencia
temperatura_input.view()
carga_input.view()
potencia_input.view()
```

```python
# Definir las reglas difusas
rule1 = ctrl.Rule(temperatura_input['alta'] & carga_input['alta'], potencia_input['baja'])
rule2 = ctrl.Rule(temperatura_input['alta'] & carga_input['media'], potencia_input['baja'])
rule3 = ctrl.Rule(temperatura_input['alta'] & carga_input['baja'], potencia_input['media'])
rule4 = ctrl.Rule(temperatura_input['media'] & carga_input['alta'], potencia_input['media'])
rule5 = ctrl.Rule(temperatura_input['media'] & carga_input['media'], potencia_input['media'])
rule6 = ctrl.Rule(temperatura_input['media'] & carga_input['baja'], potencia_input['alta'])
rule7 = ctrl.Rule(temperatura_input['baja'] & carga_input['alta'], potencia_input['alta'])
rule8 = ctrl.Rule(temperatura_input['baja'] & carga_input['media'], potencia_input['alta'])
rule9 = ctrl.Rule(temperatura_input['baja'] & carga_input['baja'], potencia_input['alta'])
```

```python
# Crear el sistema de control difuso
sistema_horno = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6, rule7, rule8, rule9])

# Simulador:
horno_prediccion = ctrl.ControlSystemSimulation(sistema_horno)
```

```python
horno_prediccion.input['temperatura'] = 250
horno_prediccion.input['carga'] = 80
horno_prediccion.compute()
print(f"Porcentaje de potencia de calentamiento: {horno_prediccion.output['potencia']:.2f} %")
```

```python
# Visualizacin grfica de la prediccin
temperatura_input.view(sim=horno_prediccion)
carga_input.view(sim=horno_prediccion)
potencia_input.view(sim=horno_prediccion)
```

```python
seg_simulacion = 20

# Partimos de la temperatura actual de 250C y una carga del 80%
temperatura_actual = 250

temperaturas = np.zeros(seg_simulacion)
potencias = np.zeros(seg_simulacion)
tiempos = np.arange(0, seg_simulacion, 1)

for t in tiempos:
    # Asignamos valores a las entradas
    horno_prediccion.input['temperatura'] = temperatura_actual

    # Ejecutamos .compute() para que haga los clculos
    horno_prediccion.compute()

    # Obtenemos las salidas
    potencia_actual = horno_prediccion.output['potencia']

    # Guardamos los datos para mostrarlos posteriormente
    temperaturas[t] = temperatura_actual
    potencias[t] = potencia_actual

    temperatura_actual = temperatura_actual + (potencia_actual * 5) / 100 - 2
```

```python
fig, ax1 = plt.subplots()

ax1.set_xlabel('Tiempo (s)')
ax1.set_ylabel('Temperatura (C)', color='tab:blue')
ax1.plot(tiempos, temperaturas, color='tab:blue')
ax1.tick_params(axis='y', labelcolor='tab:blue')

ax2 = ax1.twinx()
ax2.set_ylabel('Potencia Aplicada (%)', color='tab:red')
ax2.plot(tiempos, potencias, color='tab:red')
ax2.tick_params(axis='y', labelcolor='tab:red')

fig.tight_layout()
plt.title('Temperatura y Potencia a lo largo de 20 segundos')
plt.show()
```
