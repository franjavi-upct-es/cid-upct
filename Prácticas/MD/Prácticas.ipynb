Calcular $N=(A+B+C+D+E+F)\cdot 1000$. Con este valor $N$ ejecutar el siguiente programa:

Con este valor de $x$, definimos $X_1=10^{20}\cdot x$ y $p$ como el primer número primo mayor que $X_1$

Sean $x_0$ la primera cifra de mayor valor de $p, x_1$ la siguiente, y así consecuentemente $x_2,x_3,x_4$ y $x_5$. Es decir, $p=x_0x_1x_2x_3x_4x_5\dots$ Se construyen la matriz cuadrada simétrica $M=(m_{ij})$ de tamaño 18 de forma $m_{ii}=0$ para todo $0\le i\le 17$ y que para $i>j$,  el elemento $m_{ij}=(x_{i\mod6}\cdot x_{j\mod6}-1)\mod2$. Dibujar el grafo definido por la matriz $M$, calcular el grado de sus vértices y determinar si el grado es o no conexo

````python
from math import exp
from sympy import nextprime
import numpy as np
import networkx as nx

A, B, C, D, E, F = 6, 4, 9, 1, 1, 0

# Calcular N
N = (A + B + C + D + E + F) * 1000

print("N:", N)

def f(x):
    return 30 * x * exp(-x)
    
x, n = 1.5, N
for i in range(n):
    x = f(x)

print("x:", x)

X1 = int(1e20 * x)
p = nextprime(X1)
print("X1:", X1)
print("p:", p)

cifras_p = [int(digito) for digito in str(p)]
print(cifras_p)

# Crear la matriz de adyacencia directamente
M = np.zeros((18, 18), dtype=int)

for i in range(18):
    for j in range(i + 1, 18):
        M[i, j] = (cifras_p[i % 6] * cifras_p[j % 6] - 1) % 2

# Reflejar la matriz sobre la diagonal principal para hacerla simétrica
M = M + M.T

print("Matriz M:")
print(M)

# Crear un grafo a partir de la matriz de adyacencia
G = nx.Graph(M)

# Visualizar el grafo
nx.draw(G, with_labels=True)

# Calcular el grado de los vértices
grados = G.degree()
print("Grado de los vértices:", dict(grados))

# Determinar si el grafo es conexo
es_conexo = nx.is_connected(G)
print("El grafo es conexo:", es_conexo)
````
